# Фотонные кристаллы. Задания

## Общие задания

1. Рассмотрим одномерный фотонный кристалл с двумя чередующимися слоями равной толщины воздуха и диэлектрика с показателем преломления, равным $2$.

	а. Постройте на одном графике зонную диаграмму бесконечного кристалла и спектр коэффициента отражения от многослойной структуры с конечным числом периодов с таким же заполнением, как у кристалла (общей осью будет ось частот). Выберете масштабы и пределы по оси частот так, чтобы в выбранном диапазоне укладывалось несколько зон кристалла. Как зависит точность расчета зон от максимального индекса в усеченном ряде Фурье? Как зависит коэффициент отражения от четности числа слоев выбора верхнего первого слоя в многослойной структуре?

	б. Выпишите в явном виде френелевский коэффициент отражения на границе однородной изотропной среды и одноосного однородного кристалла, ось которого перпендикулярна границе. Запишите S-матрицу плоского слоя такого кристалла. Рассмотрите приближение сплошной среды для одномерного фотонного кристалла и с помощью расчетов методом S-матриц постройте зависимость ошибки приближения от соотношения $\lambda/\Lambda$ ($\lambda$ -- вакуумная длина волны, $\Lambda$ -- период структуры) для коэффициента отражения от многослойной структуры с конечным числом периодически чередующихся слоёв, рассматривая такую структуру сначала строго, а потом приближенно как однородный анизотропный слой.

	в. Постройте дисперсионную диаграмму со всеми разрешенными состояниями для одномерного фотонного кристалла, период которого состоит из двух диэлектрических слоёв, в координатах $\omega,\;\beta$, где $\beta$ -- постоянная распространения мод вдоль слоев кристалла.

2. Постройте зонную диаграмму в приближении пустой решетки для трехмерного кристалла с гранецентрированной кубической решеткой.
3. Рассмотрите двумерный фотонный кристалл с квадратной решеткой и одним цилиндром на периоде, и расчет его зонной диаграммы с помощью метода плоских волн (см. файл PhC2D_band_diagram.py). Охарактеризуйте сходимость метода при увеличении числа используемых плоских волн. Постройте графики зависимости числа числа плоских волн, необходимых для получения точности не хуже 1%, от контраста показателя преломления в кристалле и номера исследуемой зоны.
4. Рассмотрим брэгговское зеркало, состоящее из $N$ пар четвертьволновых диэлектрических пластин с показателями преломления $n_{1,2}$ и толщинами $l_{1,2}$. Получите рекуррентную формулу, зависящую от числа пар пластин, для вычисления волнового импеданса зеркала. С помощью полученной формулы выразите амплитудный коэффициент отражения зеркала.

## Инженерный трек

5. Рассмотрите брэгговскую решетку с конечным числом слов, состоящую из четвертьволновых пар пластинок нитрида галлия и нитрида алюминия. Постройте зависимость энергетического коэффициента отражения от частоты падающей волны. Что изменится, если одну из пластинок в каждой паре заменить на полуволновую?
6. Спроектируйте брэгговское зеркало минимальной толщины, состоящее из пар пластинок с показателями преломления $n_{1} = 1.6$, $n_{2} = 3.6$, расположенные между средами с показателями преломления $n_{b} = 1.46$, $n_{a} = 1$ для основной длины волны $\lambda_0 = 1.15\;мкм$, так чтобы в полосе $\Delta\lambda = 0.2\;мкм$ около основной длины волны коэффициент отражения был не менее $0.98$.

## Вычислительный трек

5. Рассмотрите двумерные фотонные кристаллы с квадратной и гексагональной ячейкой, состоящие либо из высокопреломляющих цилиндров в воздухе, либо из воздушных цилиндрических каналах в высокопреломляющем диэлектрике. Постройте карты фотонных запрещенных зон - двумерные области существования запрещенных зон в координатах $(\omega\Lambda/2\pi c,\;r/\Lambda)$, - и найдите параметры, при которых существуют пересекающиеся запрещенные зоны для обеих поляризаций.
6. Напишите код для расчета зонных диаграмм в двумерных кристаллах с дисперсией и исследуйте, как влияет слабая дисперсия в диэлектриках на ширину запрещенной зоны.

## Теоретический трек

5. С помощью теоремы Гельмана-Фейнмана получите формулы для вычисления групповых скоростей мод двумерного фотонного кристалла в рамках Фурье-метода.
6. Рассмотрите двумерный фотонный кристалл с квадратной решеткой для в случае ТМ поляризации. Какова степень вырождения состояния в точке $M$ первой зоны Бриллюэна, какие комбинации векторов обратной решетки соответствуют этому состоянию? Найдите приближенное аналитическое решение задачи на собственные значения в точке $M$ для слабого контраста диэлектрической проницаемости. Нарисуйте качественно расщепление состояний на дисперсионной диаграмме, соответствующее рассмотренному случаю. Как будет влиять асимметрия включения на дисперсионную диаграмму в окрестности этой точки?
