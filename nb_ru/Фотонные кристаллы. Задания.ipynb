# Фотонные кристаллы. Задания

## Общие задания

1. Рассмотрите двумерный фотонный кристалл с квадратной решеткой и одним цилиндром на периоде, и расчет его зонной диаграммы с помощью метода плоских волн. Охарактеризуйте сходимость метода при увеличении числа используемых плоских волн. Постройте графики зависимости числа числа плоских волн, необходимых для получения точности не хуже 1%, от контраста показателя преломления в кристалле и номера исследуемой зоны.
2. Рассмотрим брэгговское зеркало, состоящее из $N$ пар четвертьволновых диэлектрических пластин с показателями преломления $n_{1,2}$ и толщинами $l_{1,2}$. Получите рекуррентную формулу, зависящую от числа пар пластин, для вычисления волнового импеданса зеркала. С помощью полученной формулы выразите амплитудный коэффициент отражения зеркала.

## Инженерный трек

1. Рассмотрите брэгговскую решетку, состоящую из четвертьволновых пар пластинок. Постройте зависимость энергетического коэффициента отражения от частоты падающей волны. Что изменится, если один из слоев в каждой паре заменить одну из пластинок заменить на полуволновую?
2. Спроектируйте брэгговское зеркало минимальной толщины, состоящее из пар пластинок с показателями преломления $n_{1} = 1.6$, $n_{2} = 3.6$, расположенные между средами с показателями преломления $n_{b} = 1.46$, $n_{a} = 1$ для основной длины волны $\lambda_0 = 1.15\;мкм$, так чтобы в полосе $\Delta\lambda = 0.2\;мкм$ около основной длины волны коэффициент отражения был не менее $0.98$.

## Вычислительный трек

1. Рассмотрите двумерные фотонные кристаллы с квадратной и гексагональной ячейкой, состоящие либо из высокопреломлеющих цилиндров в воздухе, либо из воздушных цилиндрических каналах в высокопреломляющем диэлектрике. Постройте карты фотонных запрещенных зон - двумерные области существования запрещенных зон в координатах $(\omega\Lambda/2\pi c,\;r/\Lambda)$, - и найдите параметры, при которых существуют пересекающиеся запрещенные зоны для обеих поляризаций.
2. Напишите код для расчета зонных диаграмм в двумерных кристаллах с дисперсией и исследуйте, как влияет слабая дисперсия в диэлектриках на ширину запрещенной зоны.

## Теоретический трек

1. С помощью теоремы Гельмана-Фейнмана получите формулы для вычисления групповых скоростей мод фотонного кристалла с рамках Фурье-метода.
2. Рассмотрите двумерный фотонный кристалл с квадратной ячейкой и слабым контрастом диэлектрической проницаемости. Найдите приближенное аналитическое решение задачи на собственные значения в точке $M$. Как будет влиять асимметрия включения на дисперсионную диаграмму?
