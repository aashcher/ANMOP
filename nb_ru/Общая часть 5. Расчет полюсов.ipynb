# Лекция 1.5. Расчет полюсов мероморфных функций.

## Резонансные свойства и комплексные полюса

На лекции, посвященной резонансным состояниям, было получено представление матрицы рассеяния в форме разложения по резонансным состояниям
\begin{equation}\tag{1}
     S(\omega) = S^b(\omega) - \sum_n \frac{S^p_{n}(\omega)}{\omega-\omega_n^p}
\end{equation}
Далее будем пренебрегать зависимостью фоновой части $S^b$ и амплитуд $S^p_n$ от частоты. Численный расчет полюсов $\omega_n^p$ и вычетов $S^p_n$ может быть направлен либо на расчет полюсов комплекснозначитных матричных или скалярных (отдельных элементов матриц рассеяния) функций, либо на расчет нулей соответствующих обратных матриц или их элементов. Ниже рассмотрим различные подходы, которые нашли свое применение при анализе оптических и фотонных систем.

## Метод Ньютона

Классический метод Ньютона поиска нулей некоторой функции $f(x)$ основан на разложении в ряд Тейлора вблизи предполагаемого нуля $x_0$ $ f(x) = f(x_0 + \Delta x) \approx f(x_0) + \Delta x f'(x_0) $. Отсюда можно сформулировать итеративную схему метода
\begin{equation}\tag{2}
    x_{k+1} = x_k - f(x_{k})/f'(x_k).
\end{equation}
В случае комплекснозначной функции необходимо учитывать, что множество начальных приближений, из которых итеративный процесс сходится к заданному нулю, вообще говоря, является фрактальным, и поиск всех полюсов на заданном частотном интервале может оказаться нетрививиальной задачей.

В случае матричной функции запишем разложение в окрестности полюса, аналогичное разложению скалярной функции:
\begin{equation*}
    S^{-1}(\omega) \approx S^{-1}(\omega_n^p) + (\omega-\omega_n^p) \left[ \frac{dS^{-1}}{d\omega} \right]_{\omega_n^p}.
\end{equation*}
Умножим матрицы справа на некоторый вектор амплитуд $\boldsymbol{a}$ и запишем итерационный процесс через решение задачи на собственные значения
\begin{equation*}
    (\omega_i - \omega_{i+1}) \left[ \frac{dS^{-1}}{d\omega} \right]_{\omega_i} \boldsymbol{a} = S^{-1}(\omega_i) \boldsymbol{a},
\end{equation*}
откуда
\begin{equation}\tag{3}
    \omega_{i+1} = \omega_i - \min \mathrm{eig} \left(S^{-1}(\omega_i) , \left[ \frac{dS^{-1}}{d\omega} \right]_{\omega_i} \right).
\end{equation}
Здесь матрица рассеяния может вычислена различными методами, некоторые из которых рассматриваются в данном курсе. Также оказывается удобным свести расчет производной обратной матрицы рассеяния к производной самой матрицы с помощью следующего преобразования:
\begin{equation*}
    S^{-1}S = \mathbb{I} \Rightarrow \frac{dS^{-1}}{d\omega} S + S^{-1} \frac{dS}{d\omega} = 0 \Rightarrow \frac{dS^{-1}}{d\omega} = -S^{-1} S' S^{-1}
\end{equation*}
При расчете производной с помощью конечной разности необходимо следить за тем, чтобы интервал дифференцирования был достаточно малым, чтобы не оказывать влияния на итеративный процесс.

## Методы, основанные на второй производной

С помощью следующего члена разложения в ряд Тейлора можно построить метод более высокого порядка. Из разложения вблизи нуля $f(x_0+\Delta x) \approx f(x_0) + f'(x_0)\Delta x + \frac{1}{2}f''(x_0)(\Delta x)^2$ построим итерационный процесс
\begin{equation}\tag{4}
    x_{k+1} = x_k - \frac{f'(k_k)}{f''(x_k)} \left( 1 - \sqrt{1 - 2\frac{f''(x_k)f(x_k)}{[f'(x_k)]^2}} \right) \approx x_k - \frac{f(k_k)}{f'(x_k)} \left( 1 + \frac{1}{2} \frac{f''(x_k)f(x_k)}{[f'(x_k)]^2} \right)
\end{equation}

Для матричных функций вернемся к резонансному разложению матрицы рассеяния (1) и предположим, что в некотором рассматриваемом частотном диапазоне существенный вклад в матрицу оказывают конечное небольшое число слагаемых. Учтем то, что амплитуда в каждом полюсе $S_n^p$ описывает разложение некоторого резонансного состояния в выбранном базисе векторных волновых функций, и является матрицей малого ранга $r_n$, поэтому может быть разложена на произведение двух прямоугольных матриц, которые мы обозначим как $L_n$ и $R_n$, имеющих размеры $N\times r_n$ и $r_n\times N$ соответственно, где $N$ - размер S-матрицы. Тогда
\begin{equation}\tag{5}
    S(\omega) \approx S^b + \sum_n L_n \frac{1}{\omega-\omega_n^p} R_n = S^b + L(\omega\mathbb{I} - \Omega^p)^{-1}R,
\end{equation}
где $\Omega^p = \mathrm{diag}(\omega_n^p)$, а размеры прямоугольных матриц $L$ и $R$ соответствуют рангу приближенной матрицы рассеяния $r=\sum r_n$, который, вообще говоря, может быть ниже ее размерности. Чтобы найти собственные частоты два раза продифференцируем матрицу
\begin{equation}\tag{6}
    \begin{array}{c}
    S'(\omega) = -L (\omega\mathbb{I} - \Omega^p)^{-2} R \\
    S''(\omega) = 2 L (\omega\mathbb{I} - \Omega^p)^{-3} R
    \end{array}
\end{equation}
и воспользуемся сингулярным разложением второй производной $S'' = U \Sigma V^{\dagger}$, где $U$ и $V$ - прямоугольные матрицы размера $N\times r$, такие что $U^{\dagger}U = V^{\dagger}V = \mathbb{I}$, $UU^{\dagger} = VV^{\dagger} = \mathbb{I}$, а $\Sigma$ - диагональная матрица сингулярных значений. Выразим из системы (6) диагональную матрицу $\omega\mathbb{I} - \Omega^p$:
\begin{equation*}
    \begin{array}{c}
    U^{\dagger} S'(\omega) V = - U^{\dagger}L (\omega\mathbb{I} - \Omega^p)^{-2} RV \\
    \Sigma = 2 U^{\dagger}L (\omega\mathbb{I} - \Omega^p)^{-3} RV
    \end{array}
    \Rightarrow
    \begin{array}{c}
    (\omega\mathbb{I} - \Omega^p)^{-2} = - (U^{\dagger}L)^{-1} U^{\dagger} S'(\omega) V (RV)^{-1} \\
    (\omega\mathbb{I} - \Omega^p)^{-3} = (U^{\dagger}L)^{-1} \Sigma (RV)^{-1}
    \end{array}
\end{equation*}

\begin{equation*}
    \omega\mathbb{I} - \Omega^p = (\omega\mathbb{I} - \Omega^p)^{-2} (\omega\mathbb{I} - \Omega^p)^3 = -2 (U^{\dagger}L)^{-1} U^{\dagger} S'(\omega) V \Sigma^{-1} (U^{\dagger}L) \Rightarrow
\end{equation*}

\begin{equation*}
    \Rightarrow U^{\dagger} S'(\omega) V \Sigma^{-1} = \frac{1}{2} (U^{\dagger}L) (\Omega^p - \omega\mathbb{I})  (U^{\dagger}L)^{-1}
\end{equation*}
Последнее выражение представляет собой разложение по собственным значениям, так итерационный процесс запишется как
\begin{equation}\tag{7}
    \omega_{i+1} = \omega_{i} + 2 \min \mathrm{eig} \left[ U_i^{\dagger} S'(\omega_i) V_i \Sigma_i^{-1} \right]
\end{equation}
В случае, когда рассматривается всего один полюс, и $r=1$, можно показать, что последнее выражение переходит в
\begin{equation}\tag{8}
    \omega_{i+1} = \omega_{i} + 2 \frac{\max \mathrm{eig} \left[S'(\omega_i)\right]}{\max \mathrm{eig} \left[S''(\omega_i)\right]}
\end{equation}

## Методы, основанные на интеграле Коши

Контурный интеграл от комплекснозначной функции $f(z)$
\begin{equation}\tag{9}
    \oint_{C} f(z) dz = 0
\end{equation}
если $C$ - замкнутный контур, лежащий внутри области, где $f(z)$ регулярна (голоморфна). Это утверждение называется теоремой Коши. На ее основании можно доказать основную теорему о вычетах
\begin{equation}\tag{10}
    \oint_{\tilde{C}} f(z) dz = 2\pi i\sum_k \underset{z_k}{\mathrm{res}} f(z)
\end{equation}
где $\tilde{C}$ - замкнутый контур, охватывающий область, где функция является мероморфной. Из последнего утверждения следует теорема о числе нулей и полюсов (принцип аргумента)
\begin{equation}\tag{11}
    \frac{1}{2\pi i} \oint_{\tilde{C}} \frac{f'(z)}{f(z)} dz = \frac{1}{2\pi} \Delta_{\tilde{C}} \mathrm{arg} f = N - P
\end{equation}
где $N$ и $P$ - полное число нулей и полюсов, в области, охватываемой контуром.

Был предложен ряд методов, основанных на формулах (10) и (11) для вычисления положений и амплитуд полюсов. Рассмотрим здесь один из них, предложенный в работе [2]. Он основан на том факте, что при обходе вокруг изолированного нуля или полюса порядка $N$ функция меняет фазу на $2\pi N$. На первом шаге алгоритма производится триангуляция рассматриваемой области комплесной плоскости (например, на равносторонние треугольники). На втором шаге во всех узлах триангуляции вычисляются значения функции $f(z_n)$ и точки распределяются по значениям четырех квадрантам фазвой плоскости, в которых оказывается вычисленные фазы. Затем ищутся все стороны треугольников, концы которых сооветствуют изменению квадранта фазы более, чем на 1. Затем вблизи этих сторон происходит уточнение дискретизации и шаг алгоритма повторяется.

Рассматрим и другой подход на примере действий с матрицей рассеяния целиком. Запишем интегралы от матрицы рассеяния и воспользуемся теоремой о вычетах (10):
\begin{equation}\tag{12}
    I_1 = \frac{1}{2\pi i} \oint_{\Gamma} S(\omega) d\omega = S_r = \sum_k S_{r,k} = LR
\end{equation}

\begin{equation}\tag{13}
    I_2 = \frac{1}{2\pi i} \oint_{\Gamma} \omega S(\omega) d\omega = \sum_k \omega_k^p S_{r,k} = L \Omega^p R 
\end{equation}
Значения матриц $C_{1,2}$ могут быть вычислены напрямую с помощью численного интегрирования по контуру в комплексной плоскости. Действуя по аналогии с подходом, примененным для вывода уравнения (7), можно выразить матрицу $\Omega^p$:
\begin{equation}\tag{14}
    \Omega^p = \mathrm{diag} \; \mathrm{eig} (U^{\dagger}_C C_2 V_C \Sigma^{-1}_C)
\end{equation}
где $C_1 = U_C \Sigma_C V_C^{\dagger}$ - сингулярное разложение матрицы (12).

## Расчет полюсов в помощью полиномиальной интерполяции

Рассмотрим скалярную мероморфную функцию, которую, например, представляет собой один из элементов матрицы рассеяния. Согласно теореме Миттаг-Леффлера,
\begin{equation}\tag{15}
    f(\omega) = \sum_{j=1}^m \frac{a_j}{\omega - \omega_j^p} + \sum_{k=0}^{k=\infty} b_k \omega^k
\end{equation}
Введем многочлен, корни которого являются полюсами $f(\omega)$ в рассматриваемой области:
\begin{equation}\tag{16}
    P_m(\omega) = \prod_{j=1}^m (\omega-\omega_j^p) = \sum_{k=0}^m p_k \omega^k
\end{equation}
По определению, произведение $g(\omega) = P_m(\omega)f(\omega)$ - регулярная в рассматриваемой области функция. Путь ряд Тейлора этой функции сходится. Тогда производные функции $g(\omega)$ должны убывать. Предположим, что $n+1$-я производная приблизительно равна нулю. Её можно приблизить с помощью разделенных разностей Ньютона, так что
\begin{equation*}
    \sum_{i=0}^n \frac{g(\omega)}{\prod_{j\neq i} (\omega_i-\omega_j)} = \sum_{k=0}^m p_k \sum_{i=0}^n \frac{f_i \omega_i^k}{\prod_{j\neq i} (\omega_i-\omega_j)} \approx 0
\end{equation*}
где $f_i = f(\omega_i)$. Точки $\omega_i$, $i=0,1,\dots,n$ представляют собой набор точек в рассматриваемой области. Далее мы будем предполагать, что область поиска полюсов пересекает действительную ось и эти точки лежат на действительной оси. Поскольку $p_m = 1$ по определению, то последнее уравнение представляет собой алгебраическое уравнение на неизвестные коэффициенты $p_k$. Выбирая еще $m-2$ набора точек $\omega_i$, можно получить систему из $m-1$ линейных уравнений на эти коэффициенты, решая которую их легко вычислить. Полюса $\omega^p_i$ получаются из $p_k$ с помощью стандартного подхода через составление сопровождающей матрицы и решение задачи на собственные значения. После того, как найдены полюса, можно еще раз воспользоваться явным представлением $g(\omega)$ и её интерполяцией с помощью многочлена Лагранжа, куда подставляя значения $\omega = \omega^p_i$, можно вычислить амплитуды полюсов:
\begin{equation*}
    g(\omega) = \left( \sum_{j=1}^m \frac{a_j}{\omega - \omega_j^p} + \sum_{k=0}^{k=\infty} b_k \omega^k \right) \prod_{j=1}^m (\omega-\omega_j^p) =
\end{equation*}

\begin{equation*}
    = \sum_{i=0}^n f_i P_m(\omega_i) \frac{\prod_{j\neq i}(\omega-\omega_j)}{\prod_{j\neq i}(\omega_i-\omega_j)} \Rightarrow
\end{equation*}

\begin{equation}\tag{17}
    \Rightarrow a_k = \frac{\prod_{j=0}^n (\omega_k^p-\omega_j)}{\prod_{l\neq k}(\omega^p_k-\omega^p_l)} \sum_{i=0}^n f_i \frac{\prod_{l=1}^m (\omega_i-\omega_l^p)}{(\omega_k^p-\omega_i) \prod_{j\neq i}(\omega_i-\omega_j)}
\end{equation}
