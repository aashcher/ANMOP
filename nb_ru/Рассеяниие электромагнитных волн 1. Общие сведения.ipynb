# Лекция 4.1. Рассеяние электромагнитных волн

## Введение

Процессы рассеяния электромагнитного излучения возникают при изучении огромного разнообразия физических объектов начиная от планетарной и атмосферной оптики и заканчивая физикой элементарных частиц. Здесь мы ограничимся обсуждением классического упругого рассеяния электромагнитных волн, для которого постановка задачи включает рассмотрение рассеивателя - двумерного или трехмерного объекта, занимающего область двумерного или трехмерного пространства $\mathcal{D}$, объем которого $V_s$ целиком находится внутри некоторой конечной сферы радиуса $R_{out}$ c центом в выбранной декартовой системе координат $OXYZ$ (или $OXY$ в двумерном случае). В бесконечной области однородного изотропного пространства $\mathbb{R}^3\backslash\mathcal{D}$.

Строгое с точки зрения решений уравнений Максвелла рассмотрение необходимо в случаях, когда характерный размер рассеивателя/рассеивателей, или характерный размер пространственных неоднородностей в задаче сравним с длиной волны. Для малых размеров используют либо приближение эффеквной среды, либо приближение рэлеевского рассеяния. Для больших объектов применяют приближение физической оптики, геометрическую теорию дифракции или даже приближение геометрической оптики. С точки зрения как математических, так и численных методов анализа рассеяния, когда необходимо применять строгую электромагнитную теорию, различают случаи одночастичного и многочастичного рассеяния. Ниже они обсуждаются отдельно.

## Амплитуда рассеяния

Падающее на который рассеиватель поле заданных внешних источников в однородном изотропном пространстве может быть разложено по плоским волнам, поэтому рассмотрим плоскую падающую монохроматическую волну в форме $\boldsymbol{E}^{inc}(\boldsymbol{r}) = \hat{\boldsymbol{e}}_{inc} E_{inc} e^{i\boldsymbol{k}\boldsymbol{r}}$. Рассеянное поле в дальней зоне излучения представляет собой сферическую волну, поле которой запишем в виде
\begin{equation}\tag{1}
    \boldsymbol{E}^{sca}(\boldsymbol{r}) \sim \hat{\boldsymbol{e}}_{sca} f\left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right) E_{inc} \frac{e^{ikr}}{r}
\end{equation}
Функция $f\left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right)$ называется **амплитудой рассеяния**.
Для описания рассеянного поля в дальней зоне, как правило, имеет значение изменение поляризации падающей волны и зависимость комлексной амплитуды рассеяния от относительного направления векторов $\hat{\boldsymbol{k}}_{inc}$ и $\hat{\boldsymbol{k}}_{sca}$.

При разложении полей по поляризации имеет смысл рассматривать некоторый базис из двух векторов для падающей и рассеянной волны соответственно, $\hat{\boldsymbol{e}}^{p1,2,inc}$ и $\hat{\boldsymbol{e}}^{p1,2,sca}$б так что разложения полей будут записываться как
$$ \boldsymbol{E}^{inc}(\boldsymbol{r}) = (\hat{\boldsymbol{e}}^{p1,inc} E^{inc}_{1} + \hat{\boldsymbol{e}}^{p2,inc}E^{inc}_{2}) e^{i\boldsymbol{k}\boldsymbol{r}} $$

$$ \boldsymbol{E}^{sca}(\boldsymbol{r}) = (\hat{\boldsymbol{e}}^{p1,sca} E^{sca}_{1} + \hat{\boldsymbol{e}}^{p2,sca}E^{sca}_{2}) \frac{e^{ikr}}{r} $$
так что амплитуды падающего и рассеянного полей будут связаны через амплитудную матрицу рассеяния
\begin{equation}\tag{2}
    \left(\!\! \begin{array}{c} E^{sca}_{1} \\ E^{sca}_{2} \end{array} \!\!\right) = \left(\!\! \begin{array}{cc} f_{11}(\hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca}) & f_{12}(\hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca}) \\ f_{21}(\hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca}) & f_{22}(\hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca}) \end{array} \!\!\right) \left(\!\! \begin{array}{c} E^{inc}_{1} \\ E^{inc}_{2} \end{array} \!\!\right)
\end{equation}

## Сечения рассеяния, поглощения и экстинкции

Энергетические характеристики рассеяния описываются с помощью сечений рассеяния, экстинкции и поглощения. Для их определения рассмотрим усредненный по времени вектор Пойнтинга падающей плоской волны
\begin{equation}\tag{3}
	\boldsymbol{S}^{inc} = \frac{1}{2} \Re e \left( \boldsymbol{E}^{inc} \times \boldsymbol{H}^{inc\ast} \right) = \frac{|E^{inc}|^2}{2\eta}\hat{\boldsymbol{k}}_{inc}
\end{equation}
где $\eta = \sqrt{\mu/\varepsilon}$ - волновой импеданс однородной изотропной среды, в которой распространяется волна. Аналогично. для рассеянной волны
\begin{equation}\tag{4}
	\boldsymbol{S}^{sca} = \frac{|E^{sca}|^2}{2\eta}\hat{\boldsymbol{k}}_{sca} = \frac{|f\left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right)|^2}{r^2} \frac{|E^{inc}|^2}{2\eta} \hat{\boldsymbol{k}}_{sca}
\end{equation}
Тогда дифференциальная мощность, рассеиваемая в некоторый телесный угол $d\Omega_{sca}$,
\begin{equation}\tag{5}
	dP_{sca} = |\boldsymbol{S}^{sca}| r^2 d\Omega_{sca} = |f\left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right)|^2 \frac{|E^{inc}|^2}{2\eta} d\Omega_{sca}
\end{equation}
Размерность отношения 
\begin{equation}\tag{6}
	\frac{dP_{sca}}{|\boldsymbol{S}^{inc}|} = |f \left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right)|^2 d\Omega_{sca} = \sigma_{sca} \left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right) d\Omega_{sca}
\end{equation}
это площадь, а величина $\sigma_{sca}$ называется **дифференциальным сечением рассеяния**. Интегрирование по всем углам даёт **полное сечение рассеяния**
\begin{equation}\tag{7}
	\sigma_{sca} = \intop_{4\pi} \sigma_{sca} \left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right) d\Omega_{sca}
\end{equation}
Если рассматривать прокцию площади рассеивателя на плоскость, перпендикулярную направлению распространения падающей волны, $\sigma_g$, можно ввести мощность $P_g$, которая эффективно забирается рассеивающей частицей из падающей волны с точки зрения геометрической оптики
\begin{equation}\tag{8}
	P_g = \sigma_g |\boldsymbol{S}^{inc}|
\end{equation}
Если характерный размер рассеивающей частицы $D$ много меньше длины волны падающего излучения, то говорят, что реализуется режим рэлеевского рассеяния, и
\begin{equation}\tag{9}
	\frac{\sigma_s}{\sigma_g} \sim \left(\frac{D}{\lambda}\right)^4 \ll 1
\end{equation}

При наличии поглощения в материале частицы часть мощности падающей волны уходит в поглощение:
\begin{equation}\tag{10}
	P_{abs} = \frac{\omega}{2} \intop_V d\boldsymbol{r} \Im m\{\varepsilon_s\} |\boldsymbol{E}_{int}(\boldsymbol{r})|^2
\end{equation}
где $\boldsymbol{E}_{int}$ - поле внутри частицы. Тогда cечение поглощения
\begin{equation}\tag{11}
	\sigma_{abs} = \frac{P_{abs}}{|\boldsymbol{S}_{inc}|}
\end{equation}
и полное сечение или сечение экстинкции
\begin{equation}\tag{12}
	\sigma_{ext} = \sigma_{sca} + \sigma_{abs}
\end{equation}


Сечение экстинкции можно также рассматривать через взаимодействие падающего и рассеянного полей. Рассматрим некоторую точку наблюдения вне сферы, описанной вокруг рассеивающей частицы. В силу суперпозиции поле в этой точке есть сумма падаюего и рассеянного:
$$ \boldsymbol{E}(\boldsymbol{r}) = \boldsymbol{E}^{inc}(\boldsymbol{r})+ \boldsymbol{E}^{sca}(\boldsymbol{r}) $$
$$ \boldsymbol{H}(\boldsymbol{r}) = \boldsymbol{H}^{inc}(\boldsymbol{r})+ \boldsymbol{H}^{sca}(\boldsymbol{r}) $$
Тогда усредненный по времени вектор Пойнтинга в этой точке можно представить в виде суперпозиции вектора Пойнтинга падащей волны, рассеянной волны, и вектора - результата их взамодействия.
\begin{align*}
	\boldsymbol{S}(\boldsymbol{r}) = & \frac{1}{2} \Re e |\boldsymbol{E}(\boldsymbol{r}) \times \boldsymbol{H}^{\ast}(\boldsymbol{r}) | = \\
	= & \frac{1}{2} \Re e |\boldsymbol{E}^{inc}(\boldsymbol{r}) \times \boldsymbol{H}^{inc\ast}(\boldsymbol{r}) | + \frac{1}{2} \Re e |\boldsymbol{E}^{sca}(\boldsymbol{r}) \times \boldsymbol{H}^{sca\ast}(\boldsymbol{r}) | + \\
	& + \frac{1}{2} \Re e |\boldsymbol{E}^{inc}(\boldsymbol{r}) \times \boldsymbol{H}^{sca\ast}(\boldsymbol{r}) + \boldsymbol{E}^{sca}(\boldsymbol{r}) \times \boldsymbol{H}^{inc\ast}(\boldsymbol{r}) | = \\
	= & \boldsymbol{S}^{inc} + \boldsymbol{S}^{sca} + \boldsymbol{S}^{ext}
\end{align*}
Пусть $\mathcal{S}$ - некоторая сфера большого радиуса, так что все ее точки лежат в дальней зоне относительно частицы. Поскольку окружающее частицу пространство предполагается непоглощающим, поглощаемую мощность можно найти как интеграл по поверхности этой сферы:
\begin{equation}\tag{13}
	P_{abs} = -\oint_{\mathcal{S}} d\mathcal{S} \boldsymbol{S}(\boldsymbol{r})\cdot \hat{\boldsymbol{r}} = -r^2 \int_{4\pi} d\Omega \boldsymbol{S}(\boldsymbol{r})\cdot \hat{\boldsymbol{r}}
\end{equation}
Обозначим мощности падающего, рассеянного излучения и мощность экстинкции как
\begin{equation}\tag{14}
	P_{inc} = -r^2 \int_{4\pi} d\Omega \boldsymbol{S}^{inc}(\boldsymbol{r})\cdot \hat{\boldsymbol{r}} = 0
\end{equation}

\begin{equation}\tag{15}
	P_{sca} = r^2 \int_{4\pi} d\Omega \boldsymbol{S}^{sca}(\boldsymbol{r})\cdot \hat{\boldsymbol{r}}
\end{equation}

\begin{equation}\tag{16}
	P_{ext} = -r^2 \int_{4\pi} d\Omega \boldsymbol{S}^{ext}(\boldsymbol{r})\cdot \hat{\boldsymbol{r}}
\end{equation}
Отсюда получаем
\begin{equation}\tag{17}
	P_{ext} = P_{sca} + P_{abs}
\end{equation}

При падении плоской волны сечения экстинкции и рассеяния можно вразить через амплитуды полей соответствующих плоских падающей и рассеянной волн:
\begin{equation}\tag{18}
	\sigma_{ext} = \frac{P_{ext}}{|\boldsymbol{S}^{inc}|} = \frac{4\pi}{k} \hat{\boldsymbol{e}}_{sca}\cdot\hat{\boldsymbol{e}}_{inc} \Im m \{ f(\hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{inc}) \}
\end{equation}

\begin{equation}\tag{19}
	\sigma_{sca} = \frac{1}{|E^{inc}|^2} \intop_{4\pi} |\boldsymbol{S}^{sca}|^2 d\Omega
\end{equation}
где использовано разложение
$$ e^{i\boldsymbol{k}\cdot\boldsymbol{r}} = \frac{2\pi i}{k} \left[ \frac{e^{-ikr}}{r} \delta(\hat{\boldsymbol{k}}_{inc}+\hat{\boldsymbol{r}}) - \frac{e^{ikr}}{r} \delta(\hat{\boldsymbol{k}}_{inc}-\hat{\boldsymbol{r}}) \right] $$
Формула (18) представляет собой **оптическую теорему**.

## Рэлеевское рассеяние

О рэлеевском рассеянии говорят, когда размер частицы много меньше длины волны. В этом случае поле ее излучения хорошо аппроксимируется полем диполя:
\begin{equation}\tag{20}
	\boldsymbol{E}^{sca} = -\frac{k^2 e^{ikr}}{4\pi\varepsilon r} \hat{\boldsymbol{k}^{sca}} \times (\hat{\boldsymbol{k}^{sca}} \times \boldsymbol{p})
\end{equation}
где $\boldsymbol{p}$ - амплитуда индуцированного внешним полем дипольного момента. Эту амплитуду можно найти из связи между вектором поляризации и полем внутри частицы
$$ \boldsymbol{P} = (\varepsilon_s - \varepsilon) \boldsymbol{E}^{int} $$
и предположения об однородности внутреннего поля в объеме частицы $V_s$. Отсюда
$$ \boldsymbol{p} = V_s (\varepsilon_s - \varepsilon) \boldsymbol{E}^{int} $$
и
\begin{equation}\tag{20}
	\boldsymbol{E}^{sca} = -\frac{k^2 e^{ikr}}{4\pi\varepsilon r} V_s (\varepsilon_s - \varepsilon) \hat{\boldsymbol{k}^{sca}} \times (\hat{\boldsymbol{k}^{sca}} \times \boldsymbol{E}^{int}) = \frac{k^2 e^{ikr}}{4\pi\varepsilon r} V_s (\varepsilon_s - \varepsilon) \left[ \hat{\boldsymbol{e}}^{p1} (\hat{\boldsymbol{e}}^{p1} \cdot \boldsymbol{E}^{int}) + \hat{\boldsymbol{e}}^{p2} (\hat{\boldsymbol{e}}^{p2} \cdot \boldsymbol{E}^{int}) \right]
\end{equation}

В случае сферической частицы радиуса $R$ и объемом $V_s = (4\pi/3)R^3$ внутреннее и падающе поля связаны простым соотношением
$$ \boldsymbol{E}^{int} = \frac{3\varepsilon}{\varepsilon_s + 2\varepsilon} \boldsymbol{E}^{inc} $$
что позволяет записать рассеянное поле как
\begin{equation}\tag{21}
	\boldsymbol{E}^{sca} = \frac{e^{ikr}}{r} f_0 \left[ \hat{\boldsymbol{e}}^{p1} E^{inc,1} + \hat{\boldsymbol{e}}^{p2} E^{inc,2} \right],\;f_0 = k^2 R^3 \frac{\varepsilon_s - \varepsilon}{\varepsilon_s + 2\varepsilon}
\end{equation}
Рассмотрим разложение волн на TE и ТМ поляризацию относительно направления распространения падающей волны. Тогда компоненты матрицы рассеяния, выраженные через углы сферической системы координат, ось $Z$ которой задана направлением распространения падающей волны, будут равны
$$ f_{he} = f_0 \cos\vartheta_{sca}\sin(\varphi_{sca}-\varphi_{inc}) $$
$$ f_{ee} = f_0 \cos(\varphi_{sca}-\varphi_{inc}) $$
Тогда для падающей ТЕ волны полное сечение рассеяния будет определяться интегралом по углам рассеяния
\begin{equation}\tag{22}
	\sigma_{sca} = \intop_0^{\pi} d\vartheta_{sca}\sin\vartheta_{sca} \intop_0^{2\pi} d\varphi_{sca} (|f_{he}|^2 + |f_{ee}|^2) = \frac{8\pi}{3}k^4 R^6 \bigg| \frac{\varepsilon_s - \varepsilon}{\varepsilon_s + 2\varepsilon} \bigg|^2
\end{equation}
