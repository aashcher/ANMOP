# Лекция 4.1. Рассеяние электромагнитных волн

## Введение

Процессы рассеяния электромагнитного излучения возникают при изучении огромного разнообразия физических объектов начиная от планетарной и атмосферной оптики и заканчивая физикой элементарных частиц. Здесь мы ограничимся обсуждением классического упругого рассеяния электромагнитных волн, для которого постановка задачи включает рассмотрение рассеивателя - двумерного или трехмерного объекта, занимающего область двумерного или трехмерного пространства $\mathcal{D}$, объем которого $V_s$ целиком находится внутри некоторой конечной сферы радиуса $R_{out}$ c центом в выбранной декартовой системе координат $OXYZ$ (или $OXY$ в двумерном случае). В бесконечной области однородного изотропного пространства $\mathbb{R}^3\backslash\mathcal{D}$.

Строгое с точки зрения решений уравнений Максвелла рассмотрение необходимо в случаях, когда характерный размер рассеивателя/рассеивателей, или характерный размер пространственных неоднородностей в задаче сравним с длиной волны. Для малых размеров используют либо приближение эффеквной среды, либо приближение рэлеевского рассеяния. Для больших объектов применяют приближение физической оптики, геометрическую теорию дифракции или даже приближение геометрической оптики. С точки зрения как математических, так и численных методов анализа рассеяния, когда необходимо применять строгую электромагнитную теорию, различают случаи одночастичного и многочастичного рассеяния. Ниже они обсуждаются отдельно.

## Основные параметры рассеяния

Падающее на который рассеиватель поле заданных внешних источников в однородном изотропном пространстве может быть разложено по плоским волнам, поэтому рассмотрим плоскую падающую монохроматическую волну в форме $\boldsymbol{E}^{inc}(\boldsymbol{r}) = \hat{\boldsymbol{e}}_{inc} E_{inc} e^{i\boldsymbol{k}\boldsymbol{r}}$. Рассеянное поле в дальней зоне излучения представляет собой сферическую волну, поле которой запишем в виде
\begin{equation*}
    \boldsymbol{E}^{sca}(\boldsymbol{r}) \sim \hat{\boldsymbol{e}}_{sca} f\left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right) E_{inc} \frac{e^{ikr}}{r}
\end{equation*}
Функция $f\left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right)$ называется амплитудой рассеяния.
Для описания рассеянного поля в дальней зоне, как правило, имеет значение изменение поляризации падающей волны и зависимость комлексной амплитуды рассеяния от относительного направления векторов $\hat{\boldsymbol{k}}_{inc}$ и $\hat{\boldsymbol{k}}_{sca}$.
