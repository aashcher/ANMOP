# Лекция 4.1. Рассеяние электромагнитных волн

## Общие сведения

Процессы рассеяния электромагнитного излучения возникают при изучении огромного разнообразия физических объектов начиная от планетарной и атмосферной оптики и заканчивая физикой элементарных частиц. Здесь мы ограничимся обсуждением классического упругого рассеяния, для которого постановка задачи включает рассмотрение рассеивателя - двумерного или трехмерного объекта, занимающего область двумерного или трехмерного пространства $\mathcal{D}$, объем которого $V_s$ целиком находится внутри некоторой конечной сферы радиуса $R_{out}$ c центом в выбранной декартовой системе координат $OXYZ$ (или $OXY$ в двумерном случае).
