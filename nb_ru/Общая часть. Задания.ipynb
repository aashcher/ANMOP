## Инженерный и вычислительный треки

### Общие задания

1. Из уравнений Максвелла получите закон сохранения импульса для анизотропной среды с электрическими источниками и зарядами, выраженный через максвелловский тензор напряжений и плотности импульса поля $\boldsymbol{G} = \boldsymbol{D}\times\boldsymbol{B}$
2. Получите выражение для проекции вектора Пойнтинга в однородном плоском слое на ось, перпендикулярную этому слою, через амплитуды разложения поля в слое по плоским волнам.
3. Получите соотношение для эффективных амплитуд плоского источника в области этого источника, окруженного планарными структурами с известными матрицами рассеяния $S^{(1,2)}$.
4. Покажите, что векторные волны $\boldsymbol{M}$ и $\boldsymbol{N}$ удовлетворяют векторному уравнению Гельмгольца при условии, что скалярная функция, из которой они образованы, удовлетворяет скалярному уравнению Гельмгольца.
5. Найдите приближения для действительной и мнимой частей показателя преломления слабо поглощающего диэлектрика с действительной проводимостью $\sigma$ и мнимой частью диэлектрической проницаемости $\varepsilon''$, где $\varepsilon=\varepsilon'+i(\varepsilon''+\sigma/\omega)$. В морской воде $\sigma \approx 4. Сименс/м$, $\varepsilon'=81\varepsilon_0$, $\varepsilon''=0$. Чему равен импеданс морской воды? Постройте график зависимости глубины проникновения радиоволн в воде от их частоты в диапазоне от $1\;кГц$ до $1\;ГГц$.
6. Рассмотрим плоскую линейно поляризованную волну, распространяющуюся в вакууме вдоль направления оси $z$ декартовой системы координат. Пусть $Z(z) = E(z)/\eta_0H(z)$ и $Y(z)=1/Z(z)$ - нормированные волновые сопротивление и проводимость в точке $z$. Здесь $\eta_0=\sqrt{\mu_0/\varepsilon_0}$. Как связаны эти величины в некоторых точках $z_1$ и $z_2$, таких что $z_2-z_1=\ell$?
7. Постройте карту коэффициента отражения плоской волны от плоского диэлектрического слоя в координатах длина волны - проекция волнового вектора волны на плоскость слоя.

### Инженерный трек

8. Рассмотрим применение метода матриц рассеяния и переноса для анализа возбуждения с помощью призмы волноводных мод в толстом слое нематического жидкого кристалла. Данные по структуре жидкокристаллической ячейки приведены в статье https://doi.org/10.1080/02678298708086641. Воспроизведите графики, соответствующие возбуждению поверхностного плазмонного резонанса и волноводных мод, приведенные в этой статье. В каких случаях метод матриц переноса приводит к неверным результатам?
9. Простейшим антиотражающим покрытием является четвертьволновая пластинка. Покажите, что в случае нормального падения ее показатель преломления должен удовлетворять условию $n=\sqrt{n_an_b}$, $n_{a,b}$ - показатели преломления сред, окружающих пластинку. Оцените долю энергетических потерь на отражение, если толщина пластинки будет отличаться от расчетной на 1%.

## Теоретический трек

1. Получите из соотношений Теллегена условия на материальные тензоры для Лоренц-взаимной среды.
2. Запишите в явном виде в декартовых координатах "магнитную" функцию Грина, то есть, функцию Грина, которая позволяет получить электрическое поле магнитных источников, или магнитное поле электрических источников.
3. Покажите соответствие между асимптотиками векторных сферических волн на бесконечности и ТЕ/ТМ поляризованными плоскими волнами.
4. Получите коэффициенты отражения для векторных сферических волн на идеально проводящей сферической границе.
5. Получите выражение для радиальной проекции вектора Пойнтинга в однородном изотропном сферическом слое через амплитуы разложения поля в слое по сферическим волнам.
6. Получите явный вид тензора $\mathcal{L}$ в декартовых координатах при вычитании бесконечно малого кубического объема, содержащего сингулярную точку в объемном интегральном уравнении для электрического поля:
$$ \lim_{V_{\delta}\rightarrow0} \dfrac{1}{4\pi k_{0}^{2}} \oint_{S_{\delta}} \dfrac{\hat{n}\hat{e}_{R'}}{R'^{2}}\boldsymbol{J}\left(\boldsymbol{r}'\right)dS' = \dfrac{1}{k_{0}^{2}} \mathcal{L}\boldsymbol{J}\left(\boldsymbol{r}\right)$$
7. Проинтегрируйте по $k_z$ при условии, что $k_0=const$
$$ \iiint_{-\infty}^{\infty} d^{3} \boldsymbol{k}e^{i\boldsymbol{k}\left(\boldsymbol{r}-\boldsymbol{r}'\right)}\left[\dfrac{\mathbb{I}k_{0}^{2}-\boldsymbol{k}\boldsymbol{k}}{k_{0}^{2}\left(k^{2}-k_{0}^{2}\right)}+\dfrac{\hat{z}\hat{z}}{k_{0}^{2}}\right] $$
8. Рассмотрите плоский неоднородный слой в области $-h/2\leq z\leq h/2$. Чему равны амплитуды плоских волн в плоскости $z=z_1$, возбуждаемых источником, расположенным в плоскости $z=z_2$, если $-h/2\leq z_{1,2}\leq h/2$? Матрицы рассеяния между заданными слоями и границами слоя считать известными. Пользуясь эти результатом, запишите тензорную функцию Грина, разложенную по плоским волнам, для произвольной планарной многослойной структуры, состоящей из конечного числа однородных изотропных слоёв.
