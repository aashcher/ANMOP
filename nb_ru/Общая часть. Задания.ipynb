## Инженерный и вычислительный треки

### Общие задания

1. Из уравнений Максвелла получите закон сохранения импульса для анизотропной среды с электрическими источниками и зарядами, выраженный через максвелловский тензор напряжений и плотности импульса поля $\boldsymbol{G} = \boldsymbol{D}\times\boldsymbol{B}$
2. Пользуясь разложением поля в области вне источников по плоским волнам
\begin{equation*}
    \left( \begin{array}{c} \boldsymbol{E}\left(\boldsymbol{r}\right) \\ \boldsymbol{H}\left(\boldsymbol{r} \right) \end{array} \right)
    = \iint_{-\infty}^{\infty} dk_{x}dk_{y} e^{i\boldsymbol{\rho}\boldsymbol{\varkappa}} \sum_{\sigma=\pm} \left( \! \begin{array}{c} a^{s\sigma}_{\boldsymbol{k}} \hat{\boldsymbol{e}}^{s\sigma}_{\boldsymbol{k}} - \dfrac{k}{\omega\varepsilon} a^{p\sigma}_{\boldsymbol{k}} \hat{\boldsymbol{e}}^{p\sigma}_{\boldsymbol{k}}  \\  a^{p\sigma}_{\boldsymbol{k}} \hat{\boldsymbol{e}}^{s\sigma}_{\boldsymbol{k}} + \dfrac{k}{\omega\mu} a^{s\sigma}_{\boldsymbol{k}} \hat{\boldsymbol{e}}^{p\sigma}_{\boldsymbol{k}} \end{array} \! \right) e^{\sigma ik_{z}z}
\end{equation*}
получите выражение для проекции вектора Пойнтинга на ось $Z$, через амплитуды разложения поля в слое по плоским волнам.
3. Пусть известны амплитуды разложения поля излучения некоторого плоского источника по векторным плоским волнам $c^{\pm}_{\boldsymbol{\kappa}}$, когда этот источник располагается в однородном изотропном пространстве. Получите выражение для амплитуд разложения полного поля в области этого источника в случае, когда он окружен с двух сторон планарными структурами с известными матрицами рассеяния $S^{(1,2)}$.
4. Покажите, что векторные волны $\boldsymbol{M}$ и $\boldsymbol{N}$ удовлетворяют векторному уравнению Гельмгольца при условии, что скалярная функция, из которой они образованы, удовлетворяет скалярному уравнению Гельмгольца.
5. Рассмотрим плоскую линейно поляризованную волну, распространяющуюся в вакууме вдоль направления оси $z$ декартовой системы координат. Пусть $Z(z) = E(z)/\eta_0H(z)$ и $Y(z)=1/Z(z)$ - нормированные волновые сопротивление и проводимость в точке $z$. Здесь $\eta_0=\sqrt{\mu_0/\varepsilon_0}$. Как связаны эти величины в некоторых точках $z_1$ и $z_2$, таких что $z_2-z_1=\ell$?
6. Постройте карту коэффициента отражения плоской волны от плоского диэлектрического слоя с показателем преломления $n=1.5$, находящегося в воздухе, в координатах длина волны - проекция волнового вектора волны на плоскость слоя для всего возможного диапазона углов падения и длин волн от десятых долей толщины слоя до нескольких толщин.

### Инженерный трек

7. Простейшим антиотражающим покрытием является четвертьволновая пластинка. Покажите, что в случае нормального падения ее показатель преломления должен удовлетворять условию $n=\sqrt{n_an_b}$, $n_{a,b}$ - показатели преломления сред, окружающих пластинку. Аналитически оцените долю энергетических потерь на отражение, если толщина пластинки будет отличаться от расчетной на 1%. Постройте зависимость энергии, теряемой на отражении, при небольшом отклонении угла падения от нормали.
8. Рассмотрите приближенный расчет коэффициентов отражения и прохождения плоской волны через плоский слой, расположенный в воздухе, с линейно меняющимся показателем преломления от начального значения $n=1.5$ до конечного $\alpha n$, применяя метод S-матриц (файл stmatrix.py содержит необхохдимые функции). Какое число разбиений необходимо для получения точности не хуже $0.1\%$? Рассмотрите различные значения $\alpha$ и различные соотношения между толщиной слоя и длиной волны. Как зависит точность расчета от угла падения и поляризации волны? Результаты представьте в виде графиков/таблиц на ваше усмотрение.
9. Рассмотрите расчет квазинормальных мод и функции Грина для плоской пластинки. Скачайте код из дополнительного материала к статье Philip Trøst Kristensen, Kathrin Herrmann, Francesco Intravaia, and Kurt Busch, [Modeling electromagnetic resonators using quasinormal modes](https://opg.optica.org/aop/fulltext.cfm?uri=aop-12-3-612&id=434464), Adv. Opt. Photon. 12, 612-708 (2020). Исследуйте аппроксимацию функции Грина с помощью квазинормальных мод, изменяя число мод в разложении. Воспроизведите Рисунок 9 из статьи. Продемонстрируйте расходимость разложения при $x\rightarrow\infty$.

## Теоретический трек

1. Получите условия на материальные тензоры в соотношении Теллегена для Лоренц-взаимной среды.
2. Электрическая функция Грина векторного уравнения Гельмгольца в однородном изотропном пространстве с условиями излучения на бесконечности представляет собой поле диполя. Запишите в явном виде в декартовых координатах "магнитную" функцию Грина, то есть, функцию Грина, которая позволяет получить электрическое поле магнитных источников, или магнитное поле электрических источников.
3. Получите коэффициенты отражения для векторных сферических волн на идеально проводящей сферической границе.
4. Продемонстрируйте соответствие между матрицами рассеяния сферического слоя и плоского волновода, записанных в базисах сферических волн и плоских волн соответственно, при стремлении радиуса слоя к бесконечности.
5. Пусть поле в однородной изотропной области пространства свободной от источников можно разложить по сферическим волнам
\begin{align*}
    \boldsymbol{E}(k\boldsymbol{r}) &= \sum_{m,n}\sum_{\sigma=1,3} \left[ a^{\sigma}_{mn} \boldsymbol{M}^{\sigma}_{mn}(k\boldsymbol{r}) + b^{\sigma}_{mn} \boldsymbol{N}^{\sigma}_{mn}(k\boldsymbol{r}) \right]
\end{align*}
Получите выражение для радиальной проекции вектора Пойнтинга через амплитуы разложения поля.
6. Получите явный вид тензора $\mathcal{L}$ в декартовых координатах при вычитании бесконечно малого кубического объема, содержащего сингулярную точку в объемном интегральном уравнении для электрического поля:
$$ \lim_{V_{\delta}\rightarrow0} \dfrac{1}{4\pi k_{0}^{2}} \oint_{S_{\delta}} \dfrac{\hat{n}\hat{e}_{R'}}{R'^{2}}\boldsymbol{J}\left(\boldsymbol{r}'\right)dS' = \dfrac{1}{k_{0}^{2}} \mathcal{L}\boldsymbol{J}\left(\boldsymbol{r}\right)$$
7. Проинтегрируйте по $k_z$ при условии, что $k_0=const$
$$ \iiint_{-\infty}^{\infty} d^{3} \boldsymbol{k}e^{i\boldsymbol{k}\left(\boldsymbol{r}-\boldsymbol{r}'\right)}\left[\dfrac{\mathbb{I}k_{0}^{2}-\boldsymbol{k}\boldsymbol{k}}{k_{0}^{2}\left(k^{2}-k_{0}^{2}\right)}+\dfrac{\hat{z}\hat{z}}{k_{0}^{2}}\right] $$

Матрицы рассеяния между любыми заданными плоскостями считать известными. Пользуясь этим результатом, предложите метод расчета матричных элементов функции Грина, разложенной по плоским волнам, для s-поляризации, для произвольной планарной многослойной структуры, состоящей из конечного числа однородных изотропных слоёв.
