# Lecture 4.3. Electromagnetic scattering. Waterman T-matrix and the Extended Boundary Condition Method.

## Waterman T-matrix

The analysis of single-particle scattering is often conveniently carried out using the vector spherical wave expansions introduced in [Lecture 1.1](). In a given spherical coordinate system with its center lying inside the scattering volume, for example coinciding with the center of the scattering sphere in the case of Mie scattering, the external field, which may be a plane wave or a Gaussian beam, is expanded solely into regular wave functions, since the field is finite at the origin:
\begin{equation}\tag{1}
    \boldsymbol{E}^{ext}(k\boldsymbol{r}) = \sum_{m,n} \left[ c^{ext,h}_{mn} \boldsymbol{M}^{(1)}_{mn}(k\boldsymbol{r}) + c^{ext,e}_{mn} \boldsymbol{N}^{(1)}_{mn}(k\boldsymbol{r}) \right]
\end{equation}
The scattered field is represented in the form of an expansion in waves singular at zero, determined through the spherical Hankel functions of the first kind, and has the form of diverging spherical waves at large distances from the scatterer:
\begin{equation}\tag{2}
    \boldsymbol{E}^{sca}(k\boldsymbol{r}) = \sum_{m,n} \left[ a^{h}_{mn} \boldsymbol{M}^{(3)}_{mn}(k\boldsymbol{r}) + a^{e}_{mn} \boldsymbol{N}^{(3)}_{mn}(k\boldsymbol{r}) \right]
\end{equation}
Then the Waterman T-matrix, by definition, is the matrix relating the amplitude vectors $\boldsymbol{c}^{inc} = \{ c^{ext,h,e}_{mn} \}$ and $\boldsymbol{a} = \{ a^{h,e}_{mn} \}$
\begin{equation}\tag{3}
    \boldsymbol{a} = T \boldsymbol{c}^{ext}
\end{equation}

The T-matrix can be explicitly expressed through the scattering matrix. To do this, it is necessary to note, firstly, that the scattering matrix connects the vector of amplitudes of the "incoming" waves, expressed through the Hankel functions of the second kind $h^{(2)}(z)$, corresponding to the expansion of the field
\begin{equation}\tag{4}
    \boldsymbol{E}^{inc}(k\boldsymbol{r}) = \sum_{m,n} \left[ c^{inc,h}_{mn} \boldsymbol{M}^{(2)}_{mn}(k\boldsymbol{r}) + c^{inc,e}_{mn} \boldsymbol{N}^{(2)}_{mn}(k\boldsymbol{r}) \right]
\end{equation}
so that $\boldsymbol{a} = S\boldsymbol{c}^{inc}$. Further, taking into account the relationship between the spherical functions $2j(z) = h^{(1)}(z) + h^{(2)}(z)$ and the linearity of the vector spherical harmonics in the spherical functions included in them, we obtain
$$ S = \mathbb{I} + 2T $$
The unitarity condition of the scattering matrix for particles without absorption $SS^{\dagger} = \mathbb{I}$ leads to the corresponding formula for the T-matrix: $T + T^{\dagger} + 2TT^{\dagger} = 0$.

The Waterman T-matrix can be analytically averaged over the orientations of the scattering particle, so that the averaged extinction and scattering cross sections are written as
$$ \langle C_{ext} \rangle -\frac{2\pi}{k_m^2} \Re e \sum_{mn} (T^{11}_{mn,mn} + T^{22}_{mn,mn}) $$
$$ \langle C_{sca} \rangle \frac{2\pi}{k_m^2} \Re e \sum_{mn} \sum_{\alpha,\beta=1,2} |T^{\alpha\beta}_{mn,mn}|^2 $$

## Extended boundary conditions

The condition for the relation between the fields inside and outside the scattering particle can be obtained using Green's vector theorem:
\begin{equation}\tag{5}
    \intop_V dV \left[ \boldsymbol{a}\cdot(\nabla\times\nabla\times\boldsymbol{b}) - \boldsymbol{b}\cdot(\nabla\times\nabla\times\boldsymbol{a}) \right] = \intop_{\partial V} dS \hat{\boldsymbol{n}} \cdot \left[ \boldsymbol{b}\times(\nabla\times\boldsymbol{a}) - \boldsymbol{a}\times(\nabla\times\boldsymbol{c}) \right]
\end{equation}
where $\hat{\boldsymbol{n}}$ is the normal to the surface $\partial V$ bounding the volume $V$. We apply equation (5) to the region $V_{ext}$ external to the scattering particle, which is a sphere of sufficiently large radius with its center inside the scatterer, $V\in V_{ext}\in \mathbb{R}^3$, and substitute $\boldsymbol{a}=\boldsymbol{E}$ and $\boldsymbol{b}=\mathcal{G}_0(\boldsymbol{r},\boldsymbol{r}')\boldsymbol{c}$, where $\boldsymbol{c}$ is some constant vector. Using the Helmholtz equation, it is easy to obtain
\begin{equation}\tag{6}
    \begin{cases}  \boldsymbol{E}(\boldsymbol{r}')\cdot\boldsymbol{c}, & \boldsymbol{r}'\in V_{ext} \\ 0, & \boldsymbol{r}'\in V  \end{cases} = \left( -\intop_{\partial V_{ext}} + \intop_{\partial V} \right) dS \hat{\boldsymbol{n}} \cdot \left[ (\nabla\times\boldsymbol{E})\times(\mathcal{G}_0(\boldsymbol{r},\boldsymbol{r}')\boldsymbol{c}) - \boldsymbol{E}\times(\nabla\times\mathcal{G}_0(\boldsymbol{r},\boldsymbol{r}')\boldsymbol{c}) \right]
\end{equation}
The scattered field decreases with distance from the scatterer, therefore, tending the size of the region $V_{ext}$ to infinity, one can see that the first integral over the surface of this region corresponds to the external field, while the second one corresponds to the scattered field, so that the first line in equation (6) is a record of the superposition principle for the total field in the external region $\boldsymbol{E} = \boldsymbol{E}^{ext} + \boldsymbol{E}^{sca}$, where
\begin{equation}\tag{7}
    \boldsymbol{E}^{sca}(\boldsymbol{r}') = \intop_{\partial V} dS \left[ i\omega\mu_0\mathcal{G}_0(\boldsymbol{r},\boldsymbol{r}')\cdot(\hat{\boldsymbol{n}}\times\boldsymbol{H})_+ + (\nabla\times\mathcal{G}_0(\boldsymbol{r},\boldsymbol{r}'))\cdot(\hat{\boldsymbol{n}}\times\boldsymbol{E})_+ \right],\;\boldsymbol{r}'\in V_{ext}
\end{equation}
Here the arbitrariness of the constant vector $\boldsymbol{c}$ was taken into account, and Faraday's law was used to express the circulation of the electric field through the magnetic field. The index "+" indicates that the tangential component of the field is taken on the outer part of the surface of the scattering particle.

Similarly, the second line of equation (6) gives an expression for the external field inside the scattering particle:
\begin{equation}\tag{8}
    \boldsymbol{E}^{ext}(\boldsymbol{r}') = - \intop_{\partial V} dS \left[ i\omega\mu_0\mathcal{G}_0(\boldsymbol{r},\boldsymbol{r}')\cdot(\hat{\boldsymbol{n}}\times\boldsymbol{H})_+ + (\nabla\times\mathcal{G}_0(\boldsymbol{r},\boldsymbol{r}'))\cdot(\hat{\boldsymbol{n}}\times\boldsymbol{E})_+ \right],\;\boldsymbol{r}'\in V
\end{equation}
This equation is sometimes called an extended boundary condition, since it relates the field at the outer boundary of the scatterer to the external field inside the scatterer. The necessity of using such a boundary condition instead of the usual conditions on the continuity of the tangential components of the fields is due to the fact that it is difficult to prove the completeness and convergence of the expansions of the field in vector wave functions in cases of geometrically complex boundaries between different media defining the surface of the scatterers.

## Calculation of the scatterer's T-matrix

Let us apply the obtained equations to calculate the T-matrix. The idea of ​​the method is to find the tangential components of the fields on the outer surface of the particle using equation (7) based on the known external field, and then calculate the coefficients of the expansion of the scattered field into spherical waves using equation (8).

Let us use explicit expansions of the fields and the tensor Green's function outside the source region in terms of vector spherical waves:
\begin{equation}\tag{9}
    \mathcal{G}_0(\boldsymbol{r},\boldsymbol{r}') = ik_m \sum_{m,n} \begin{cases} \boldsymbol{M}^{(1)}_{mn}(k_m\boldsymbol{r}) (\boldsymbol{M}^{(3)}_{-mn}(k_m\boldsymbol{r}'))^T + \boldsymbol{N}^{(1)}_{mn}(k_m\boldsymbol{r}) (\boldsymbol{N}^{(3)}_{-mn}(k_m\boldsymbol{r}'))^T & r < r' \\ \boldsymbol{M}^{(3)}_{mn}(k_m\boldsymbol{r}) (\boldsymbol{M}^{(1)}_{-mn}(k_m\boldsymbol{r}'))^T + \boldsymbol{N}^{(3)}_{mn}(k_m\boldsymbol{r}) (\boldsymbol{N}^{(1)}_{-mn}(k_m\boldsymbol{r}'))^T & r > r' \end{cases}
\end{equation}
Let us introduce the radii of the inscribed $r_{in}$ and circumscribed $r_{out}$ spheres around the particle. Then, inside the inscribed sphere for $r'<r$ in a homogeneous spherical region of space for the external field, the expansion is valid
\begin{equation}\tag{10}
    \boldsymbol{E}^{ext}(k_p\boldsymbol{r}') = \sum_{m,n} \left[ c^{ext,h}_{mn} \boldsymbol{M}^{(1)}_{mn}(k_m\boldsymbol{r}') + c^{ext,e}_{mn} \boldsymbol{N}^{(1)}_{mn}(k_m\boldsymbol{r}') \right],\;r' < r_{in}
\end{equation}
Substituting (9) and (10) into equation (8) yields expressions for the amplitudes of the expansion of the external field:
\begin{equation}\tag{11}
    c^{ext,h}_{mn} = k_m \intop_{\partial V} dS \left[ \omega\mu_0\boldsymbol{M}^{(3)}_{-mn}(k_m\boldsymbol{r})\cdot(\hat{\boldsymbol{n}}\times\boldsymbol{H})_+ - ik_m\boldsymbol{N}^{(3)}_{-mn}(k_m\boldsymbol{r})\cdot(\hat{\boldsymbol{n}}\times\boldsymbol{E})_+ \right]
\end{equation}

\begin{equation}\tag{12}
    c^{ext,e}_{mn} = k_m \intop_{\partial V} dS \left[ \omega\mu_0\boldsymbol{N}^{(3)}_{-mn}(k_m\boldsymbol{r})\cdot(\hat{\boldsymbol{n}}\times\boldsymbol{H})_+ - ik_m\boldsymbol{M}^{(3)}_{-mn}(k_m\boldsymbol{r})\cdot(\hat{\boldsymbol{n}}\times\boldsymbol{E})_+ \right]
\end{equation}
Here, the orthogonality of vector wave functions and the formulas for transforming $\boldsymbol{M}$ into $\boldsymbol{N}$ and back using the rotor were also used.

Similarly, for $r'>r_{out}$, the substitution of the expansion of the scattered field
\begin{equation}\tag{13}
    \boldsymbol{E}^{sca}(k_p\boldsymbol{r}') = \sum_{m,n} \left[ a^{h}_{mn} \boldsymbol{M}^{(3)}_{mn}(k_p\boldsymbol{r}') + a^{e}_{mn} \boldsymbol{N}^{(3)}_{mn}(k_p\boldsymbol{r}') \right],\;r' > r_{out}
\end{equation}
gives the amplitudes
\begin{equation}\tag{14}
    a^{h}_{mn} = -k_m \intop_{\partial V} dS \left[ \omega\mu_0\boldsymbol{M}^{(1)}_{-mn}(k_m\boldsymbol{r})\cdot(\hat{\boldsymbol{n}}\times\boldsymbol{H})_+ - ik_m\boldsymbol{N}^{(1)}_{-mn}(k_m\boldsymbol{r})\cdot(\hat{\boldsymbol{n}}\times\boldsymbol{E})_+ \right]
\end{equation}

\begin{equation}\tag{15}
    a^{e}_{mn} = -k_m \intop_{\partial V} dS \left[ \omega\mu_0\boldsymbol{N}^{(1)}_{-mn}(k_m\boldsymbol{r})\cdot(\hat{\boldsymbol{n}}\times\boldsymbol{H})_+ - ik_m\boldsymbol{M}^{(1)}_{-mn}(k_m\boldsymbol{r})\cdot(\hat{\boldsymbol{n}}\times\boldsymbol{E})_+ \right]
\end{equation}

It can be shown that on the surface $\partial V$ the following decompositions are complete:
\begin{equation}\tag{16}
    \hat{\boldsymbol{n}}\times\boldsymbol{E}(\boldsymbol{r}') = \sum_{mn} \left[ b_{mn}^{h} \hat{\boldsymbol{n}}\times\boldsymbol{M}^{(1)}_{mn}(k_p\boldsymbol{r}') + b_{mn}^{e} \hat{\boldsymbol{n}}\times\boldsymbol{N}^{(1)}_{mn}(k_p\boldsymbol{r}') \right],\;\boldsymbol{r}'\in\partial V
\end{equation}

\begin{equation}\tag{17}
    \hat{\boldsymbol{n}}\times\boldsymbol{H}(\boldsymbol{r}') = \frac{1}{i\omega\mu_0} \sum_{mn} \left[ b_{mn}^{h} \hat{\boldsymbol{n}}\times\nabla'\times\boldsymbol{M}^{(1)}_{mn}(k_p\boldsymbol{r}') + b_{mn}^{e} \hat{\boldsymbol{n}}\times\nabla'\times\boldsymbol{N}^{(1)}_{mn}(k_p\boldsymbol{r}') \right],\;\boldsymbol{r}'\in\partial V
\end{equation}
Substituting these expansions into (11), (12) and (14), (15), and using the continuity of the tangential components of the field on the surface $\partial V$, we can obtain the required relationship between the coefficients of the external and scattered fields.

From (11), (12), (16) and (17) we find,
\begin{equation}\tag{18}
    \left(\!\! \begin{array}{c} \boldsymbol{c}^{ext,h} \\ \boldsymbol{c}^{ext,e} \end{array} \!\!\right) = Q \left(\!\! \begin{array}{c} \boldsymbol{b}^h \\ \boldsymbol{b}^e \end{array} \!\!\right) = \left(\!\! \begin{array}{cc} Q^{hh} & Q^{he} \\ Q^{eh} & Q^{ee} \end{array} \!\!\right) \left(\!\! \begin{array}{c} \boldsymbol{b}^h \\ \boldsymbol{b}^e \end{array} \!\!\right)
\end{equation}
Matrix elements in explicit form are
\begin{equation}\tag{19}
    \begin{array}{l}
        Q^{hh}_{mn,m'n'} = -ik_pk_m C^{31}_{mn,m'n'} - ik_m^2C^{13}_{mn,m'n'} \\
        Q^{he}_{mn,m'n'} = -ik_pk_m C^{11}_{mn,m'n'} - ik_m^2C^{33}_{mn,m'n'} \\
        Q^{eh}_{mn,m'n'} = -ik_pk_m C^{33}_{mn,m'n'} - ik_m^2C^{11}_{mn,m'n'} \\
        Q^{ee}_{mn,m'n'} = -ik_pk_m C^{13}_{mn,m'n'} - ik_m^2C^{31}_{mn,m'n'}
    \end{array}
\end{equation}
where
\begin{equation}\tag{20}
    \left(\!\! \begin{array}{c} C^{11}_{mn,m'n'} \\ C^{12}_{mn,m'n'} \\ C^{21}_{mn,m'n'} \\ C^{22}_{mn,m'n'} \end{array} \!\! \right) = \oint_{\partial V} dS \hat{\boldsymbol{n}} \cdot \left[ \begin{array}{c} \boldsymbol{M}^{(1)}_{m'n'}(k_p\boldsymbol{r}) \times \boldsymbol{M}^{(3)}_{-mn}(k_m\boldsymbol{r}) \\ \boldsymbol{M}^{(1)}_{m'n'}(k_p\boldsymbol{r}) \times \boldsymbol{N}^{(3)}_{-mn}(k_m\boldsymbol{r}) \\ \boldsymbol{N}^{(1)}_{m'n'}(k_p\boldsymbol{r}) \times \boldsymbol{M}^{(3)}_{-mn}(k_m\boldsymbol{r}) \\ \boldsymbol{N}^{(1)}_{m'n'}(k_p\boldsymbol{r}) \times \boldsymbol{N}^{(3)}_{-mn}(k_m\boldsymbol{r})  \end{array} \right]
\end{equation}
Similarly, substituting (16), (17) into (14), (15) gives the relationship between the amplitudes of the scattered field and the amplitudes of the expansion of the tangential field on the particle surface:
\begin{equation}\tag{21}
    \left(\!\! \begin{array}{c} \boldsymbol{a}^h \\ \boldsymbol{a}^e \end{array} \!\!\right) = \mathrm{Rg} Q \left(\!\! \begin{array}{c} \boldsymbol{b}^h \\ \boldsymbol{b}^e \end{array} \!\!\right)
\end{equation}
The matrix $\mathrm{Rg} Q$ differs from the matrix $Q$ in equation (18) in that the expressions corresponding to formulas (20) contain exclusively regular vector wave functions, denoted by the superscript "$(1)$".

Thus, the relationship between the amplitudes of the external and scattered fields is explicitly written as
\begin{equation}\tag{22}
    \left(\!\! \begin{array}{c} \boldsymbol{a}^h \\ \boldsymbol{a}^e \end{array} \!\!\right) = -(\mathrm{Rg} Q) Q^{-1} \left(\!\! \begin{array}{c} \boldsymbol{c}^{ext,h} \\ \boldsymbol{c}^{ext,e} \end{array} \!\!\right)\;\Rightarrow\;T = -(\mathrm{Rg} Q) Q^{-1}
\end{equation}
