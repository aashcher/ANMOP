# Lecture 4.1. Electromangetic scattering


## Introduction

Scattering processes of electromagnetic radiation arise in the study of a huge variety of physical objects, from planetary and atmospheric optics to particle physics. Here we restrict ourselves to discussing classical elastic scattering of electromagnetic waves, for which the problem statement includes consideration of a scatterer - a two-dimensional or three-dimensional object occupying a region of two-dimensional or three-dimensional space $\mathcal{D}$, the volume of which $V_s$ is entirely inside some finite sphere of radius $R_{out}$ with its center in the chosen Cartesian coordinate system $OXYZ$ (or $OXY$ in the two-dimensional case). In an infinite region of homogeneous isotropic space $\mathbb{R}^3\backslash\mathcal{D}$.

A rigorous consideration from the point of view of solutions of Maxwell's equations is necessary in cases where the characteristic size of the scatterer/scatterers, or the characteristic size of spatial inhomogeneities in the problem is comparable to the wavelength. For small sizes, either the effective medium approximation or the Rayleigh scattering approximation is used. For large objects, the physical optics approximation, the geometric theory of diffraction, or even the geometric optics approximation are used. From the point of view of both mathematical and numerical methods of scattering analysis, when it is necessary to apply rigorous electromagnetic theory, a distinction is made between cases of single-particle and multiparticle scattering. They are discussed separately below.

## Scattering amplitude

The incident field of the given external sources in a homogeneous isotropic space on which the scatterer is located can be expanded in plane waves, so we consider a plane incident monochromatic wave in the form $\boldsymbol{E}^{inc}(\boldsymbol{r}) = \hat{\boldsymbol{e}}_{inc} E_{inc} e^{i\boldsymbol{k}\boldsymbol{r}}$. The scattered field in the far radiation zone is a spherical wave, the field of which we write in the form
\begin{equation}\tag{1}
    \boldsymbol{E}^{sca}(\boldsymbol{r}) \sim \hat{\boldsymbol{e}}_{sca} f\left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right) E_{inc} \frac{e^{ikr}}{r}
\end{equation}
The function $f\left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right)$ is called the **scattering amplitude**.
To describe the scattered field in the far zone, as a rule, the change in the polarization of the incident wave and the dependence of the complex scattering amplitude on the relative direction of the vectors $\hat{\boldsymbol{k}}_{inc}$ and $\hat{\boldsymbol{k}}_{sca}$ are important.

When expanding fields by polarization, it makes sense to consider a certain basis of two vectors for the incident and scattered waves, respectively, $\hat{\boldsymbol{e}}^{p1,2,inc}$ and $\hat{\boldsymbol{e}}^{p1,2,sca}$, so that the expansions of the fields will be written as
$$ \boldsymbol{E}^{inc}(\boldsymbol{r}) = (\hat{\boldsymbol{e}}^{p1,inc} E^{inc}_{1} + \hat{\boldsymbol{e}}^{p2,inc}E^{inc}_{2}) e^{i\boldsymbol{k}\boldsymbol{r}} $$

$$ \boldsymbol{E}^{sca}(\boldsymbol{r}) = (\hat{\boldsymbol{e}}^{p1,sca} E^{sca}_{1} + \hat{\boldsymbol{e}}^{p2,sca}E^{sca}_{2}) \frac{e^{ikr}}{r} $$
so that the amplitudes of the incident and scattered fields will be related through the scattering amplitude matrix
\begin{equation}\tag{2}
    \left(\!\! \begin{array}{c} E^{sca}_{1} \\ E^{sca}_{2} \end{array} \!\!\right) = \left(\!\! \begin{array}{cc} f_{11}(\hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca}) & f_{12}(\hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca}) \\ f_{21}(\hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca}) & f_{22}(\hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca}) \end{array} \!\!\right) \left(\!\! \begin{array}{c} E^{inc}_{1} \\ E^{inc}_{2} \end{array} \!\!\right)
\end{equation}

## Scattering, absorption and extinction cross sections

The energy characteristics of scattering are described using scattering, extinction and absorption cross sections. To determine them, we consider the time-averaged Poynting vector of the incident plane wave
\begin{equation}\tag{3}
	\boldsymbol{S}^{inc} = \frac{1}{2} \Re e \left( \boldsymbol{E}^{inc} \times \boldsymbol{H}^{inc\ast} \right) = \frac{|E^{inc}|^2}{2\eta}\hat{\boldsymbol{k}}_{inc}
\end{equation}
where $\eta = \sqrt{\mu/\varepsilon}$ is the wave impedance of a homogeneous isotropic medium in which the wave propagates. Similarly, for a scattered wave
\begin{equation}\tag{4}
	\boldsymbol{S}^{sca} = \frac{|E^{sca}|^2}{2\eta}\hat{\boldsymbol{k}}_{sca} = \frac{|f\left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right)|^2}{r^2} \frac{|E^{inc}|^2}{2\eta} \hat{\boldsymbol{k}}_{sca}
\end{equation}
Then the differential power dissipated into some solid angle $d\Omega_{sca}$,
\begin{equation}\tag{5}
	dP_{sca} = |\boldsymbol{S}^{sca}| r^2 d\Omega_{sca} = |f\left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right)|^2 \frac{|E^{inc}|^2}{2\eta} d\Omega_{sca}
\end{equation}
Dimensionality of the ratio 
\begin{equation}\tag{6}
	\frac{dP_{sca}}{|\boldsymbol{S}^{inc}|} = |f \left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right)|^2 d\Omega_{sca} = \sigma_{sca} \left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right) d\Omega_{sca}
\end{equation}
is the area, and the quantity $\sigma_{sca}$ is called the **differential scattering cross section**. Integrating over all angles yields the **total scattering cross section**
\begin{equation}\tag{7}
	\sigma_{sca} = \intop_{4\pi} \sigma_{sca} \left( \hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{sca} \right) d\Omega_{sca}
\end{equation}
If we consider the ratio of the scatterer area to the plane perpendicular to the direction of propagation of the incident wave, $\sigma_g$, we can introduce the power $P_g$, which is effectively taken by the scattering particle from the incident wave from the point of view of geometric optics
\begin{equation}\tag{8}
	P_g = \sigma_g |\boldsymbol{S}^{inc}|
\end{equation}
If the characteristic size of the scattering particle $D$ is much smaller than the wavelength of the incident radiation, then we say that the Rayleigh scattering regime is realized, and
\begin{equation}\tag{9}
	\frac{\sigma_s}{\sigma_g} \sim \left(\frac{D}{\lambda}\right)^4 \ll 1
\end{equation}

If there is absorption in the particle material, part of the power of the incident wave goes into absorption:
\begin{equation}\tag{10}
	P_{abs} = \frac{\omega}{2} \intop_V d\boldsymbol{r} \Im m\{\varepsilon_s\} |\boldsymbol{E}_{int}(\boldsymbol{r})|^2
\end{equation}
where $\boldsymbol{E}_{int}$ is the field inside the particle. Then the absorption cross section
\begin{equation}\tag{11}
	\sigma_{abs} = \frac{P_{abs}}{|\boldsymbol{S}_{inc}|}
\end{equation}
and the total cross section or extinction cross section
\begin{equation}\tag{12}
	\sigma_{ext} = \sigma_{sca} + \sigma_{abs}
\end{equation}

The extinction cross section can also be considered through the interaction of the incident and scattered fields. Let us consider some observation point outside the sphere described around the scattering particle. Due to superposition, the field at this point is the sum of the incident and scattered fields:
$$ \boldsymbol{E}(\boldsymbol{r}) = \boldsymbol{E}^{inc}(\boldsymbol{r})+ \boldsymbol{E}^{sca}(\boldsymbol{r}) $$
$$ \boldsymbol{H}(\boldsymbol{r}) = \boldsymbol{H}^{inc}(\boldsymbol{r})+ \boldsymbol{H}^{sca}(\boldsymbol{r}) $$
Then the time-averaged Poynting vector at this point can be represented as a superposition of the Poynting vector of the incident wave, the scattered wave, and the vector resulting from their interaction.
\begin{align*}
	\boldsymbol{S}(\boldsymbol{r}) = & \frac{1}{2} \Re e |\boldsymbol{E}(\boldsymbol{r}) \times \boldsymbol{H}^{\ast}(\boldsymbol{r}) | = \\
	= & \frac{1}{2} \Re e |\boldsymbol{E}^{inc}(\boldsymbol{r}) \times \boldsymbol{H}^{inc\ast}(\boldsymbol{r}) | + \frac{1}{2} \Re e |\boldsymbol{E}^{sca}(\boldsymbol{r}) \times \boldsymbol{H}^{sca\ast}(\boldsymbol{r}) | + \\
	& + \frac{1}{2} \Re e |\boldsymbol{E}^{inc}(\boldsymbol{r}) \times \boldsymbol{H}^{sca\ast}(\boldsymbol{r}) + \boldsymbol{E}^{sca}(\boldsymbol{r}) \times \boldsymbol{H}^{inc\ast}(\boldsymbol{r}) | = \\
	= & \boldsymbol{S}^{inc} + \boldsymbol{S}^{sca} + \boldsymbol{S}^{ext}
\end{align*}
Let $\mathcal{S}$ be some sphere of large radius, so that all its points lie in the far zone relative to the particle. Since the space surrounding the particle is assumed to be non-absorbing, the absorbed power can be found as an integral over the surface of this sphere:
\begin{equation}\tag{13}
	P_{abs} = -\oint_{\mathcal{S}} d\mathcal{S} \boldsymbol{S}(\boldsymbol{r})\cdot \hat{\boldsymbol{r}} = -r^2 \int_{4\pi} d\Omega \boldsymbol{S}(\boldsymbol{r})\cdot \hat{\boldsymbol{r}}
\end{equation}
Let us denote the powers of incident, scattered radiation and the extinction power as
\begin{equation}\tag{14}
	P_{inc} = -r^2 \int_{4\pi} d\Omega \boldsymbol{S}^{inc}(\boldsymbol{r})\cdot \hat{\boldsymbol{r}} = 0
\end{equation}

\begin{equation}\tag{15}
	P_{sca} = r^2 \int_{4\pi} d\Omega \boldsymbol{S}^{sca}(\boldsymbol{r})\cdot \hat{\boldsymbol{r}}
\end{equation}

\begin{equation}\tag{16}
	P_{ext} = -r^2 \int_{4\pi} d\Omega \boldsymbol{S}^{ext}(\boldsymbol{r})\cdot \hat{\boldsymbol{r}}
\end{equation}
From this we get
\begin{equation}\tag{17}
	P_{ext} = P_{sca} + P_{abs}
\end{equation}

When a plane wave is incident, the extinction and scattering cross sections can be expressed through the amplitudes of the fields of the corresponding plane incident and scattered waves:
\begin{equation}\tag{18}
	\sigma_{ext} = \frac{P_{ext}}{|\boldsymbol{S}^{inc}|} = \frac{4\pi}{k} \hat{\boldsymbol{e}}_{sca}\cdot\hat{\boldsymbol{e}}_{inc} \Im m \{ f(\hat{\boldsymbol{k}}_{inc}, \hat{\boldsymbol{k}}_{inc}) \}
\end{equation}

\begin{equation}\tag{19}
	\sigma_{sca} = \frac{1}{|E^{inc}|^2} \intop_{4\pi} |\boldsymbol{S}^{sca}|^2 d\Omega
\end{equation}
where decomposition is used
$$ e^{i\boldsymbol{k}\cdot\boldsymbol{r}} = \frac{2\pi i}{k} \left[ \frac{e^{-ikr}}{r} \delta(\hat{\boldsymbol{k}}_{inc}+\hat{\boldsymbol{r}}) - \frac{e^{ikr}}{r} \delta(\hat{\boldsymbol{k}}_{inc}-\hat{\boldsymbol{r}}) \right] $$
Formula (18) is an **optical theorem**.

## Rayleigh scattering

Rayleigh scattering is said to occur when the particle size is much smaller than the wavelength. In this case, its radiation field is well approximated by the dipole field:
\begin{equation}\tag{20}
	\boldsymbol{E}^{sca} = -\frac{k^2 e^{ikr}}{4\pi\varepsilon r} \hat{\boldsymbol{k}^{sca}} \times (\hat{\boldsymbol{k}^{sca}} \times \boldsymbol{p})
\end{equation}
where $\boldsymbol{p}$ is the amplitude of the dipole moment induced by the external field. This amplitude can be found from the relationship between the polarization vector and the field inside the particle
$$ \boldsymbol{P} = (\varepsilon_s - \varepsilon) \boldsymbol{E}^{int} $$
and the assumptions about the homogeneity of the internal field in the volume of the particle $V_s$. Hence
$$ \boldsymbol{p} = V_s (\varepsilon_s - \varepsilon) \boldsymbol{E}^{int} $$
and
\begin{equation}\tag{20}
	\boldsymbol{E}^{sca} = -\frac{k^2 e^{ikr}}{4\pi\varepsilon r} V_s (\varepsilon_s - \varepsilon) \hat{\boldsymbol{k}^{sca}} \times (\hat{\boldsymbol{k}^{sca}} \times \boldsymbol{E}^{int}) = \frac{k^2 e^{ikr}}{4\pi\varepsilon r} V_s (\varepsilon_s - \varepsilon) \left[ \hat{\boldsymbol{e}}^{p1} (\hat{\boldsymbol{e}}^{p1} \cdot \boldsymbol{E}^{int}) + \hat{\boldsymbol{e}}^{p2} (\hat{\boldsymbol{e}}^{p2} \cdot \boldsymbol{E}^{int}) \right]
\end{equation}

In the case of a spherical particle of radius $R$ and volume $V_s = (4\pi/3)R^3$, the internal and incident fields are related by a simple relation
$$ \boldsymbol{E}^{int} = \frac{3\varepsilon}{\varepsilon_s + 2\varepsilon} \boldsymbol{E}^{inc} $$
which allows us to write the scattered field as
\begin{equation}\tag{21}
	\boldsymbol{E}^{sca} = \frac{e^{ikr}}{r} f_0 \left[ \hat{\boldsymbol{e}}^{p1} E^{inc,1} + \hat{\boldsymbol{e}}^{p2} E^{inc,2} \right],\;f_0 = k^2 R^3 \frac{\varepsilon_s - \varepsilon}{\varepsilon_s + 2\varepsilon}
\end{equation}
Let us consider the decomposition of waves into TE and TM polarization relative to the direction of propagation of the incident wave. Then the components of the scattering matrix, expressed through the angles of a spherical coordinate system, the $Z$ axis of which is determined by the direction of propagation of the incident wave, will be equal to
$$ f_{he} = f_0 \cos\vartheta_{sca}\sin(\varphi_{sca}-\varphi_{inc}) $$
$$ f_{ee} = f_0 \cos(\varphi_{sca}-\varphi_{inc}) $$
Then for the incident TE wave the total scattering cross section will be determined by the integral over the scattering angles
\begin{equation}\tag{22}
	\sigma_{sca} = \intop_0^{\pi} d\vartheta_{sca}\sin\vartheta_{sca} \intop_0^{2\pi} d\varphi_{sca} (|f_{he}|^2 + |f_{ee}|^2) = \frac{8\pi}{3}k^4 R^6 \bigg| \frac{\varepsilon_s - \varepsilon}{\varepsilon_s + 2\varepsilon} \bigg|^2
\end{equation}
