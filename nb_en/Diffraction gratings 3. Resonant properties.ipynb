# Lecture 3.3. Diffraction gratings. Resonant properties

## Resonances and scattering matrix poles

In the introduction, it was shown that the modes of photonic structures can be studied not only by solving the eigenvalue problem for the wave equation, but also by calculating the poles of the scattering matrix. The scattering matrix provides a unified way to describe a slab, allowing one to obtain both a discrete and a continuous spectrum. Let waves defined by the amplitude vector $\boldsymbol{a}_{inc}$ be incident on some optical structure, and let the scattered field be defined by the amplitude vector $\boldsymbol{a}_{sca}$. We write the relationship between them through the inverse scattering matrix as $S^{-1}\boldsymbol{a}_{sca} = \boldsymbol{a}_{inc}$. With a zero right-hand side, the solutions of this equation will be the eigensolutions $\boldsymbol{a}_{eig}$ of the Helmholtz equation without sources, that is, the eigenvalues ​​will be determined by the poles of the scattering matrix, and the eigenvectors will be determined by the residues at these poles.

The poles can be found by one of the numerical methods considered previously. If we study resonances in diffraction gratings, we can use the intermediate results of the Fourier-modal method, or we can approximately describe the resonances through the interference of Bloch modes. We will consider these approaches below.

## Grating modes

###  Mode calculation within the FMM

Within the framework of the Fourier-modal method, T-matrices of the interfaces of the photonic crystal layer with homogeneous media were obtained in explicit form:
\begin{equation}
    \left(\begin{array}{c} \boldsymbol{b}^{-}\\ \boldsymbol{b}^{+} \end{array}\right) = \left(\begin{array}{cc} T_{11}^{(1)} & T_{12}^{(1)} \\ T_{21}^{(1)} & T_{22}^{(1)} \end{array}\right) \left(\begin{array}{c} \mathcal{E}\boldsymbol{c}^{-} \\ \boldsymbol{c}^{+} \end{array}\right)
\end{equation}

\begin{equation}
    \left(\begin{array}{c} \boldsymbol{a}^{+} \\ \boldsymbol{a}^{-} \end{array}\right)=\left(\begin{array}{cc} T_{11}^{(2)} & T_{12}^{(2)} \\ T_{21}^{(2)} & T_{22}^{(2)} \end{array}\right)\left(\begin{array}{c} \mathcal{E}\boldsymbol{c}^{+}\\ \boldsymbol{c}^{-} \end{array}\right)
\end{equation}
From these equations we can obtain the reflection matrices for the modal coefficients at the layer boundaries:
\begin{equation}
    \boldsymbol{b}^{+} = T_{21}^{(1)} \mathcal{E}\boldsymbol{c}^{-} + T_{22}^{(1)} \boldsymbol{c}^{+}
    \Rightarrow \boldsymbol{c}^{+} = \left(T_{22}^{(1)}\right)^{-1} \boldsymbol{b}^{+} - \left(T_{22}^{(1)}\right)^{-1} T_{21}^{(1)} \mathcal{E}\boldsymbol{c}^{-} \Rightarrow R_c^{(1)} = - \left(T_{22}^{(1)}\right)^{-1} T_{21}^{(1)}
\end{equation}

\begin{equation}
    \boldsymbol{a}^{-} = T_{21}^{(2)} \mathcal{E}\boldsymbol{c}^{+} + T_{22}^{(2)} \boldsymbol{c}^{-} 
    \Rightarrow \boldsymbol{c}^{-} = \left(T_{22}^{(2)}\right)^{-1} \boldsymbol{a}^{-} - \left(T_{22}^{(2)}\right)^{-1} T_{21}^{(2)} \mathcal{E}\boldsymbol{c}^{+} \Rightarrow R_c^{(2)} = - \left(T_{22}^{(2)}\right)^{-1} T_{21}^{(2)}
\end{equation}
In the presence of a falling field, we obtain a condition for the existence of a proper solution:
\begin{equation}
    \left\{ \begin{array}{c} \boldsymbol{c}^{+} = R_c^{(1)} \mathcal{E}\boldsymbol{c}^{-} \\ \boldsymbol{c}^{-} = R_c^{(2)} \mathcal{E}\boldsymbol{c}^{+} \end{array} \right. \Rightarrow \det \left( \mathbb{I} - R_c^{(1)} \mathcal{E} R_c^{(2)} \mathcal{E} \right) = 0
\end{equation}
The solution of this equation gives the frequencies of the grating modes.

### Resonances as Bloch's mode interference

The modal representation also allows one to obtain a simple approximation and physically interpret high-Q resonances in dielectric gratings. Consider a symmetric structure for which $R_c^{(1)} = R_c^{(2)}$. Let the grating period and contrast be such that there is a small number of propagating Bloch modes in it for which $\beta_m^2 > 0$, $1\leq m\leq N_m$. Neglect all decaying modes and consider in more detail the reflection of modes at the grating boundaries.

In the single-mode approximation, when $N_m=1$, the equality of the determinant to zero leads to the condition
\begin{equation}
    1-r_{11}^{2}\exp\left(2i\beta_{1}h\right)=0\Rightarrow\beta_{1}h+\arg r_{11}=\pi l
\end{equation}
which shows that in this approximation the grating behaves exactly like a flat homogeneous dielectric layer. The amplitude of the outgoing zero harmonic
\begin{equation}
    a_1^+ = t_{1} c_1^+ \exp(i\beta_1 h)
\end{equation}
and $|t_{1}|\neq 0$ for anu value of the Bloch vector.

In the two-mode approximation, when $N_m=2$, there appear effects that are not observed for a homogeneous layer. First, at the Γ-point, the first mode is symmetric, and the second is antisymmetric. This means that the overlap integral of the external plane wave propagating normally to the grating is zero, and therefore $r_{12}=r_{21}=0$ as well as $r_{22}=1$ and $t_2 = 0$. The state with the excited second mode, which has an infinite quality factor at the Γ-point, is called a bound state in the continuum protected by symmetry. Second, for $k_B\neq0$, the two-mode approximation gives four equations for the mode amplitudes, which can be rewritten in the same form as the equations for the Fabry-Perot resonator:
\begin{equation}
    1-\left(r^{(12)}\right)^{2}\exp\left(2i\beta_{2}h\right)=0
\end{equation}
where
\begin{equation}
    r^{(12)}=\dfrac{r_{22}+\alpha r_{11}r_{21}r_{12}\exp\left(2i\beta_{1}h\right)}{1-\alpha r_{21}r_{12}\exp\left[i\left(\beta_{1}+\beta_{2}\right)h\right]}, \; \;  \alpha=\dfrac{1}{1-r_{11}^{2}\exp\left(2i\beta_{1}h\right)}
\end{equation}
Similarly, the amplitude of the outgoing plane wave can be expressed in terms of the effective transmittance:
\begin{equation}
    b^{t} = t_{1}a_{1}^{+} \exp\left(i\beta_{1}h\right) + t_{2}a_{2}^{+}\exp\left(i\beta_{2}h\right) = t^{(12)}a_{2}^{+}\exp\left(i\beta_{2}h\right)
\end{equation}

\begin{equation}
    t^{(12)}=t_{2}+\alpha t_{1}r_{21}\exp\left(i\beta_{1}h\right)\left[r^{(12)}\exp\left(i\beta_{2}h\right)+r_{11}\exp\left(i\beta_{1}h\right)\right]
\end{equation}
which can already be completely zeroed for some value of the Bloch vector. This situation can be interpreted as destructive interference of the mode fields in the space surrounding the grating. At the point where the effective transmittance is zero, the quality factor of this state is also equal to infinity, and the state is called an occasional or interference bound state in the continuum, or Friedrich-Wintgen BIC.

## Relation between pole and coupled-mode description

Let us consider models for describing the resonant behavior of waveguide gratings upon diffraction of a monochromatic plane wave incident on the grating at different angles. In the framework of the approach that describes the resonant properties of waveguide gratings through the poles of meromorphic functions of the projection of the wave vector in the plane of the grating, the main parameters for constructing the reflection coefficient curve include the off-resonance reflection coefficient, which can be considered constant in a certain range of variation of the argument of the resonance function near resonance, the coordinates of the pole on the complex plane, and the pole amplitude. On the other hand, there is a more physically intuitive approach to describing the resonant properties of waveguide gratings through the formalism of coupled modes. In this case, each mode is described by the propagation constant $\beta$, the radiation coefficient $\alpha$, which characterizes the energy loss of the mode due to the presence of the grating (in a flat waveguide it is zero), and the coupling coefficient of this mode with the plane wave incident on the grating $\kappa$. The excitation conditions (conservation of momentum in projection onto the plane of the grating) are described using the optogeometric parameter
\begin{equation}
    k_x  = k_x^{inc} \pm K_g = k_0 n_c \sin\theta_{inc} \pm K_g
\end{equation}
where $k_0=2\pi/\lambda$, and the sign $\pm$ corresponds to the coupling of the incident radiation with a mode that propagates either in the same direction as the incident wave, or in the opposite direction.

The value of describing resonant behavior using the coupled mode formalism is that it not only provides a physically intuitive description of the occurrence of resonance, but also allows one to solve the inverse problem of designing resonant devices at an intuitive level. The problem with this approach is that the parameters $\alpha$, $\beta$, and $\kappa$ cannot be measured directly. Let us consider how these parameters can be obtained by analyzing the dependence of the reflection coefficient on the projection of the wave vector $r(k_x)$.

The necessary condition for the excitation of a mode at collinear incidence of a plane wave using the "$-1$"-th diffraction order in k-space is written as
\begin{equation}
    \beta = -n_m k_0 = k_x^{inc} - K_g
\end{equation}
Let us denote the refractive indices of the substrate, waveguide and coating as $n_s$, $n_g$ and $n_c$, so that $n_g>n_s>n_c$, $n_g\equiv n_m$. The following components contribute to the reflection. First, this is the Fresnel reflection, which we denote by the symbol $r_0$. Second, this is the contribution $r_g$, caused by the coupled mode, which is re-emitted due to the presence of the grating. We write the dependence $r(k_x)$ using the contribution of the pole term
\begin{equation}
    r(k_x) = r_0 + \frac{a_p}{k_x - k_p}
\end{equation}
The variable $k_x$ depends on the problem parameters $\lambda$ and $\Lambda$, as well as on the angle of incidence $\theta_{inc}$, so that experimentally and by means of modeling one can directly obtain $r(k_x)$ as a function of the real variable by changing the angle of incidence.

The amplitude of the waveguide mode $C_g(x)$ satisfies the differential equation, meaning that the rate of change of the mode amplitude is due to excitation by the external field and radiation due to the grating:
\begin{equation}
    \frac{\partial C}{\partial x} = \kappa q(x) \exp[i(k_x-\beta)x] - \alpha C_g
\end{equation}
where $q(x)=1$ is the field profile of an infinite plane wave incident on an infinite grating. The difference $k_x-\beta$ shows the detuning from ideal synchronism of the mode excitation. The stationary solution of this equation in the limit of the synchronism condition
\begin{equation}
    C\left(x\right)=c\left(x\right)e^{-\alpha x}\Rightarrow\dfrac{dc}{dx}=\kappa e^{\alpha x+i\left(k_x-\beta\right)x}\Rightarrow C\left(x\right)=\frac{-i\kappa}{k_x-\beta-i\alpha}
\end{equation}
The resulting mode amplitude should be linearly related to the lattice contribution to the reflection coefficient.
\begin{equation}
    r_g(k_x) = \eta C = \frac{-i\eta\kappa}{k_x-(\beta+i\alpha)}
\end{equation}
The physical meaning of the coefficient $\eta$ is that it determines the diffraction losses of the mode in the coating. Comparing the obtained expression with the pole expansion, we have
\begin{split}
    a_p = -i\eta\kappa \\
    k_p = \beta+i\alpha
\end{split}

Let us consider a graphical interpretation of the obtained relationships on the complex plane of the reflection coefficient. When the parameter $k_x$ changes from $-\infty$ to $+\infty$, the reflection coefficient describes a circle on the complex plane. This follows from the fact that it is easy to show that
\begin{equation}
    \left( \Re e[r_g] + \frac{\Im m[a_p]}{2\alpha} \right)^2 + \left( \Im e[r_g] - \frac{\Re m[a_p]}{2\alpha} \right)^2 = \left( \frac{|a_p|}{2\alpha} \right)^2
\end{equation}
The center of this circle is the point $z_0 = ia_p/2\alpha$, and the radius is $\rho = a_p|/2\alpha$. The position of the point on the circle is determined by the argument
\begin{equation}
    \mathrm{Arg} [r_g(k_x) - z_0] = \varphi_p - \pi/2 - 2\arctan\left(\frac{k_x-\beta}{\alpha}\right)
\end{equation}
Off the resonance, the reflection coefficient $r_g$ becomes zero, so the circle crosses 0.

The total reflection coefficient is composed of the Fabry-Perot reflection and the reflection due to the waveguide grating mode, so the total reflection coefficient circle is shifted by $r_0$ to the point
\begin{equation}
    z_c = r_0 + ia_p/2\alpha
\end{equation}
If the grating parameters are chosen so that the modulus of the reflection coefficient at resonance is equal to 1, then the circle $r(k_x)$ touches the unit circle at the point $r_{max}$. In this case, obviously, the points of maximum $r_{max}$ and minimum $r_{min}$ reflection coefficients, as well as the center $z_c$ lie on a straight line passing through zero of the complex plane. The condition of resonant excitation of the mode $k_x=\beta$ gives an expression for the reflection coefficient at this point
\begin{equation*}
    r(k_x=\beta) = r_{beta} = r_0 + \frac{ia_p}{\alpha}
\end{equation*}
From this it is clear that $r_{\beta}$, $r_0$ and $z_c$ lie on the same line. Then
\begin{equation*}
    z_c = \frac{1}{2}(r_0 + r_{\beta}) = \frac{1}{2}(r_{max} + r_{min}) \Rightarrow r_{\beta} = r_{max} + r_{min} - r_0
\end{equation*}
that is, $r_{\beta}$ turns out to be expressed in terms of measurable or easily calculated quantities.

Using the obtained relations, it can be shown that the modulus of the reflection coefficient
\begin{equation*}
    |r(k_x)|^2 = \frac{|r_0|^2(k_x-\beta)^2 + |r_\beta|^2\alpha^2 + 2\alpha(k_x-\beta)\zeta}{(k_x-\beta)^2 + \alpha^2}
\end{equation*}
where $\zeta = \sqrt{(|r_{max}|^2-|r_0|^2)/(|r_0|^2-|r_{min}|^2)}$. From the condition $d(|r(k_x)|^2)/dk_x = 0$, we can find the projections of the wave vector corresponding to the maximum and minimum of reflection, from which we obtain explicit expressions for the diffraction loss coefficient $\alpha$ and the mode propagation constant $\beta$:
\begin{equation*}
    \begin{array}{l} k_{x,max}-\beta = \alpha/\zeta \\ k_{x,min}-\beta = \alpha\zeta \end{array} \Rightarrow \alpha = \zeta \frac{k_{x,max}-k_{x,min}}{1+\zeta^2}
\end{equation*}

\begin{equation*}
    \beta = \frac{k_{x,max}+k_{x,min}}{2} + \frac{k_{x,max}-k_{x,min}}{2} \frac{\zeta^2-1}{\zeta^2+1}
\end{equation*}
